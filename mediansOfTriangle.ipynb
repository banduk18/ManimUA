```python
from manim import *
config.media_width = "75%"
config.verbosity = "WARNING"

import math
```

```python
%%manim -qh -v WARNING mediansOfTriangle

class mediansOfTriangle(Scene):
    def construct(self):
        
        AB_len = 9.6*2/3
        CA_len = 7.6*2/3
        
        
        dot_A = Dot()
        dot_B = dot_A.copy().shift((UP*math.sin(47.21*DEGREES)+RIGHT*math.cos(47.21*DEGREES))*9.6*2/3)
        dot_C = dot_A.copy().shift(RIGHT*7.6*2/3)
        
        dots = VGroup(*[dot_A, dot_B, dot_C]).scale(0.75).shift(DOWN+LEFT)
        
        label_A = Tex("A").next_to(dot_A, LEFT)
        label_B = Tex("B").next_to(dot_B, LEFT)
        label_C = Tex("C").next_to(dot_C, RIGHT)
        
        
        self.play(LaggedStart(Create(dots), lag_ratio=0.3)
                 ,Write(label_A)
                 ,Write(label_B)
                 ,Write(label_C)
                 )
        
        AB = Line(dot_A, dot_B)
        BC = Line(dot_B, dot_C)
        CA = Line(dot_C, dot_A)
        
        lines = VGroup(*[AB, BC, CA])
        
        self.play(LaggedStart(Create(lines), lag_ratio=0.3))
        
        dot_N = Dot(BC.get_center())
        dot_D = Dot(CA.get_center())
        dot_M = Dot(AB.get_center())
        
        label_N = Tex("N").next_to(dot_N, RIGHT)
        label_D = Tex("D").next_to(dot_D, DOWN)
        label_M = Tex("M").next_to(dot_M, LEFT)
        
        median_dots = VGroup(*[dot_N, dot_D, dot_M])
        
        self.play(LaggedStart(Create(median_dots), lag_ratio=0.3)
                 ,Write(label_N)
                 ,Write(label_D)
                 ,Write(label_M)
                 )
        
        AN = Line(dot_A, dot_N)
        BD = Line(dot_B, dot_D)
        CM = Line(dot_C, dot_M)
        
        median_lines = VGroup(*[AN, BD, CM])
        
        self.play(LaggedStart(Create(median_lines), lag_ratio=0.3))
        self.wait()
        
        
        #Dot_O
        
        angle_NAC = Angle(CA,AN, quadrant=(-1,1))
        
        dot_O = dot_A.copy().shift((UP*math.sin(angle_NAC.get_value()) + RIGHT*math.cos(angle_NAC.get_value()))*AN.get_length()*2.04/3)
        label_O = Tex("O").next_to(dot_O, UP*2+LEFT*0.2)
        
        labels = VGroup(*[
            label_A,
            label_B,
            label_C,
            label_M,
            label_N,
            label_D,
            label_O
        ])
    
        self.play(Create(dot_O)
                 ,Write(label_O)
                 )
        self.wait()
        
        
        #Dashes BD
        
        #D1
        dot_D1 = Dot(Line(dot_A, dot_D).get_center())
        
        
        angle_BDC = Angle(Line(dot_D, dot_C), Line(dot_B, dot_D), radius = 0.4, quadrant=(1,-1))
        
        #self.add(angle_BDC)
        
        dash_D1 = Line(dot_D1.copy().shift((UP*math.sin(angle_BDC.get_value())+ RIGHT*math.cos(angle_BDC.get_value()))*.3),
                       dot_D1.copy().shift((DOWN*math.sin(angle_BDC.get_value())+ LEFT*math.cos(angle_BDC.get_value()))*.3)
                      )
        
        #D2
        dot_D2 = Dot(Line(dot_D, dot_C).get_center())
        
        dash_D2 = dash_D1.copy().move_to(dot_D2)
        
        self.play(Create(dash_D1)
                 ,Create(dash_D2)
                 )
        self.wait()
        
        #Dashes AN
        #N1
        dot_N1 = Dot(Line(dot_B, dot_N).get_center())
        
        angle_ANC = Angle(Line(dot_A, dot_N), Line(dot_N, dot_C), radius=0.4, quadrant = (1,-1))
        
        dash_N1 = Line(dot_N1.copy().shift((UP*math.sin(angle_NAC.get_value())+ RIGHT*math.cos(angle_NAC.get_value()))*.3),
                       dot_N1.copy().shift((DOWN*math.sin(angle_NAC.get_value())+ LEFT*math.cos(angle_NAC.get_value()))*.3)
                      )
        
        dashes_N1 = VGroup(*[dash_N1, dash_N1.copy()]).arrange(DOWN, buff = 0.1).move_to(dot_N1)
        
        #N2
        dot_N2 = Dot(Line(dot_N, dot_C).get_center())
        
        dashes_N2 = dashes_N1.copy().move_to(dot_N2)
        
        self.play(Create(dashes_N1)
                 ,Create(dashes_N2)
                 )
        self.wait()

        #Dashes CM
        #M1
        
        dot_M1 = Dot(Line(dot_M, dot_B).get_center())
        
        angle_ACM = Angle(Line(dot_A, dot_C), Line(dot_C, dot_M), radius = 0.4, quadrant=(-1,1), other_angle = True)
        
        
        dash_M1 = Line(dot_M1.copy().shift((UP*math.sin(angle_ACM.get_value())+ RIGHT*math.cos(angle_ACM.get_value()))*.3),
                       dot_M1.copy().shift((DOWN*math.sin(angle_ACM.get_value())+ LEFT*math.cos(angle_ACM.get_value()))*.3)
                      )
        
        dashes_M1 = VGroup(*[dash_M1, dash_M1.copy(), dash_M1.copy()]) \
            .arrange(UP+RIGHT, buff = 0) \
            .move_to(dot_M1)
        #M2
        dot_M2 = Dot(Line(dot_M, dot_A).get_center())
        
        dashes_M2 = dashes_M1.copy().move_to(dot_M2)
        
        self.play(Create(dashes_M1)
                 ,Create(dashes_M2)
                 )
        self.wait()
        
        dashes = VGroup(*[dash_D1, dash_D2, dashes_M1, dashes_M2, dashes_N1, dashes_N2])
        
        #PT1
        #ROW1
        TEXT1 = Text("Добудуємо паралелограм ABKC")
        ROW1 = VGroup(*[TEXT1]).arrange(RIGHT).scale(0.5)
        
        #ROW2
        TEXT2 = Text("За властивістю паралелограма:")
        ROW2 = VGroup(*[TEXT2]).arrange(RIGHT).scale(0.5)
        
        #ROW3
        TEXT3 = MathTex("AK^2")
        TEXT4 = MathTex("+")
        TEXT5 = MathTex("BC^2")
        TEXT6 = MathTex("=")
        TEXT7 = MathTex("2AC^2")
        TEXT8 = MathTex("+")
        TEXT9 = MathTex("2AB^2")
        ROW3 = VGroup(*[TEXT3,TEXT4,TEXT5,TEXT6,TEXT7,TEXT8,TEXT9]).arrange(RIGHT).scale(0.75)
        
        #ROW4
        TEXT10 = MathTex("AN^2 =")
        TEXT11 = MathTex(r"\frac{2AC^2 + 2AB^2 - BC^2}{4}")
        ROW4 = VGroup(*[TEXT10, TEXT11]).arrange(RIGHT).scale(0.75)
        
        #ROW5
        TEXT11 = Text("Повторивши для інших медіан \n\nотримаємо:")
        ROW5 = VGroup(*[TEXT11]).arrange(RIGHT).scale(0.5)
        
        #ROW6
        TEXT12 = MathTex(r"BD^2 = \frac{2AB^2 + 2BC^2 - AC^2}{4}")
        TEXT13 = MathTex(r"CM^2 = \frac{2AC^2 + 2BC^2 - AB^2}{4}")
        
        ROW6 = VGroup(*[TEXT12, TEXT13]).arrange(RIGHT, buff = 0.5).scale(0.75)
        
        ROWS1 = VGroup(*[ROW1,ROW2,ROW3,ROW4,ROW5,ROW6]).arrange(DOWN, buff = 0.5, aligned_edge = LEFT).shift(LEFT*1.5)
        
        #ROW1
        self.play(Write(ROW1))
        self.wait()
        
        #Dot_K
        
        dot_K = dot_B.copy().shift(RIGHT*(Line(dot_A, dot_C).get_length()))
        
        BK = Line(dot_K, dot_B)
        CK = Line(dot_K, dot_C)
        KN = Line(dot_K, dot_N)
        
        label_K = Tex("K").next_to(dot_K, DOWN)
        
        parr = VGroup(*[BK, CK, KN])
        
        self.play(Create(dot_K)
                 ,Write(label_K)
                 )
        self.wait()
        
        self.play(LaggedStart(Create(parr), lag_ratio=0.3))
        self.wait()
        
        #ROW2
        self.play(Write(ROW2))
        self.wait()
        
        #ROW3
        AK = Line(dot_A, dot_K)
        BC = Line(dot_B, dot_C)
        self.play(Indicate(VGroup(*[AK,BC])))
        self.play(FadeOut(VGroup(*[AK,BC])))
        self.wait()
                  
        self.play(ReplacementTransform(Line(dot_A, dot_K), TEXT3)
                 ,Write(TEXT4)
                 ,ReplacementTransform(Line(dot_B, dot_C), TEXT5)
                 )
        self.wait()
               
        BK = Line(dot_B, dot_K)
        AC = Line(dot_A, dot_C)
        AB = Line(dot_A, dot_B)
        CK = Line(dot_C, dot_K)
        
        self.play(Indicate(VGroup(*[BK,AC,AB,CK])))
        self.play(FadeOut(VGroup(*[BK,AC,AB,CK])))
        self.wait()
                  
        self.play(Write(TEXT6)
                 ,ReplacementTransform(VGroup(*[Line(dot_B, dot_K), Line(dot_A, dot_C)]), TEXT7)
                 ,Write(TEXT8)
                 ,ReplacementTransform(VGroup(*[Line(dot_A, dot_B), Line(dot_C, dot_K)]), TEXT9)
                 )
        self.wait()
        
        #ROW4
        self.play(Write(ROW4))
        self.wait()
        
        #ROW5
        self.play(Write(ROW5))
        self.wait()
        
        #ROW6
        self.play(Write(ROW6))
        self.wait()
        
        all_stuff = VGroup(*[dots, dot_M, dot_N, dot_D, dot_O,
                            labels, lines, median_lines, dashes])
        
        self.play(FadeOut(VGroup(*[ROWS1, parr, dot_K, label_K]))
                 ,all_stuff.animate.shift(UP*0.5+RIGHT*2.5))
        
        self.wait()
        
        #PT2
        
        #ROW1
        TEXT1 = Text("Додавши всі медіани отримаємо:")
        ROW1 = VGroup(*[TEXT1]).arrange(RIGHT).scale(0.5)
        
        #ROW2
        TEXT2 = MathTex(r"AN^2")
        TEXT2_1 = MathTex(r"+BD^2+CM^2 = ")
        TEXT2_2 = MathTex(r"\frac{2AB^2+2AC^2-BC^2}{4}")
        TEXT2_3 = MathTex(r"+\frac{2AB^2+2BC^2-AC^2}{4}+\frac{2AC^2+2BC^2-2AB^2}{4}")
        ROW2 = VGroup(*[TEXT2, TEXT2_1, TEXT2_2, TEXT2_3]).arrange(RIGHT).scale(0.5)
        
        #ROW3
        TEXT3 = Text("Домноживши на 4 та згрупувавши всі аргументи:")
        ROW3 = VGroup(*[TEXT3]).arrange(RIGHT).scale(0.5)
        
        #ROW4
        TEXT4 = MathTex(r"4AN^2+4BD^2+4CM^2=3AB^2+3AC^2+3BC^2")
        ROW4 = VGroup(*[TEXT4]).arrange(RIGHT).scale(0.65)
        
        #ROW5
        TEXT5 = Text("Виразимо одну з сторін (BC) через інші аргументи")
        ROW5 = VGroup(*[TEXT5]).arrange(RIGHT).scale(0.5)
        
        #ROW6
        TEXT6 = MathTex(r"3BC^2 = 4AN^2+4BD^2+4CM^2-3AB^2-3AC^2")
        ROW6 = VGroup(*[TEXT6]).arrange(RIGHT).scale(0.65)
        
        #ROW7
        TEXT7 = Text("Додамо ")
        TEXT7_2 = MathTex(r"1.5BC^2").scale(1.3)
        TEXT7_3 = Text("з кожної сторони")
        ROW7 = VGroup(*[TEXT7, TEXT7_2, TEXT7_3]).arrange(RIGHT).scale(0.5)
        
        #ROW8
        
        TEXT8 = MathTex(r"4.5BC^2 = 4AN^2+4BD^2+4CM^2 - ")
        TEXT8_2 = MathTex(r"(3AB^2+AC^2-1.5BC^2)")
        ROW8 = VGroup(*[TEXT8, TEXT8_2]).arrange(RIGHT).scale(0.65)
        
        #Indicate TEXT_8_2 to TEXT2_2
        
        #ROW9 
        
        TEXT9 = Text("Замінимо")
        TEXT9_2 = MathTex(r"(3AB^2+AC^2-1.5BC^2)").scale(1.3)
        TEXT9_3 = Text("на")
        TEXT9_4 = MathTex(r"6AN^2").scale(1.3)
        ROW9 = VGroup(*[TEXT9, TEXT9_2, TEXT9_3, TEXT9_4]).arrange(RIGHT).scale(0.5)
        
        ROWS2 = VGroup(*[ROW1,ROW2, ROW3,ROW4,ROW5,ROW6,ROW7,ROW8,ROW9]).arrange(DOWN,buff = 0.4, aligned_edge=LEFT).shift(LEFT*1.5+UP*0.5)
        
        #ROW1
        self.play(Write(ROW1))
        self.wait()
        
        #ROW2
        self.play(Write(ROW2))
        self.wait()
        
        #ROW3
        self.play(Write(ROW3))
        self.wait()
        
        #ROW4
        self.play(Write(ROW4))
        self.wait()
        
        #ROW5
        self.play(Write(ROW5))
        self.wait()
        
        #ROW6
        self.play(Write(ROW6))
        self.wait()
        
        #ROW7
        self.play(Write(ROW7))
        self.wait()
        
        #ROW8
        self.play(Write(ROW8))
        self.wait()
        
        self.play(Circumscribe(TEXT8_2))
        self.wait()
        self.play(Circumscribe(TEXT2_2))
        self.wait()
                  
        #ROW9
        self.play(Write(ROW9))
        self.wait(3)
        
        self.play(FadeOut(ROWS2))
        
        #PT3
        #ROW1
        TEXT1 = MathTex(r"4.5BC^2 = 4AN^2+4BD^2+4CM^2-6AN^2")
        ROW1 = VGroup(*[TEXT1]).arrange(RIGHT).scale(0.65)
        
        #ROW2
        TEXT2 = MathTex(r"BC = \frac{2}{3}\sqrt{2(BD^2+CM^2) - AN^2}")
        ROW2=VGroup(*[TEXT2]).arrange(RIGHT).scale(0.65)
        
        #ROW3
        TEXT3 = Text("Підставивши значення медіан за умовою отримаємо:")
        ROW3 = VGroup(*[TEXT3]).arrange(RIGHT).scale(0.5)
        
        #ROW4
        TEXT4 = MathTex(r"BC = \frac{2}{3}\sqrt{2(60^2+36^2) - 48^2}")
        TEXT4_2 = MathTex(r"\approx 57.69")
        ROW4 = VGroup(*[TEXT4,TEXT4_2]).arrange(RIGHT).scale(0.65)
        
        #ROW5
        TEXT5 = Text("Аналогічно з іншими сторонами:")
        ROW5 = VGroup(*[TEXT5]).arrange(RIGHT).scale(0.5)
        
        #ROW6
        TEXT6 = MathTex(r"AB = \frac{2}{3}\sqrt{2(60^2+48^2)-36^2}")
        TEXT6_2 = MathTex(r"\approx 68.35")
        ROW6 = VGroup(*[TEXT6, TEXT6_2]).arrange(RIGHT).scale(0.65)
        
        #ROW7
        TEXT7 = MathTex(r"AC = \frac{2}{3}\sqrt{2(36^2+48^2)-60^2}")
        TEXT7_2 = MathTex(r"= 40 ")
        ROW7 = VGroup(*[TEXT7,TEXT7_2]).arrange(RIGHT).scale(0.65)
        
        ROWS3 = VGroup(*[ROW1, ROW2, ROW3, ROW4, ROW5,ROW6,ROW7]).arrange(DOWN, buff = 0.4, aligned_edge=LEFT).shift(LEFT*2)
        
        #ROW1
        self.play(Write(ROW1))
        self.wait()
        
        #ROW2
        self.play(Write(ROW2))
        self.wait()
        
        #ROW3
        self.play(Write(ROW3))
        self.wait()
        
        #ROW4
        self.play(Write(TEXT4))
        self.wait()
        self.play(Write(TEXT4_2))
        self.wait()
        
        #ROW5
        self.play(Write(ROW5))
        self.wait()
        
        #ROW6
        self.play(Write(TEXT6))
        self.wait()
        self.play(Write(TEXT6_2))
        self.wait()
        
        #ROW7
        self.play(Write(TEXT7))
        self.wait()
        self.play(Write(TEXT7_2))
        self.wait(3)
        
        self.play(FadeOut(ROWS3))
        #self.remove(rows)
        
        #PT4
        #ROW1
        TEXT1 = Text("За формулою Герона знаходимо площу трикутника:")
        ROW1 = VGroup(*[TEXT1]).arrange(RIGHT).scale(0.5)
        
        #ROW2
        TEXT2 = MathTex(r"S = \sqrt{p(p-a)(p-b)(p-c)}")
        ROW2 = VGroup(*[TEXT2]).arrange(RIGHT).scale(0.65)
        
        #ROW3
        TEXT3 = MathTex(r"p = \frac{57.69 + 68.35 + 40}{2}")
        TEXT3_2 = MathTex(r" = 83.02")
        ROW3 = VGroup(*[TEXT3, TEXT3_2]).arrange(RIGHT).scale(0.65)
        
        #ROW4
        TEXT4 = MathTex(r"S = \sqrt{83.02(83.02 - 40)(83.02-57.69)(83.02 - 68.35)}")
        ROW4 = VGroup(*[TEXT4]).arrange(RIGHT).scale(0.65)
        
        #ROW5
        TEXT5 = MathTex(r"S \approx 1152.02")
        ROW5 = VGroup(*[TEXT5]).scale(0.65)
        
        ROWS = VGroup(*[ROW1,ROW2,ROW3,ROW4,ROW5]).arrange(DOWN, buff = 0.4, aligned_edge=LEFT).shift(LEFT*2+UP*1.5)
        
        #ROW1
        self.play(Write(ROW1))
        self.wait()
        
        #ROW2
        self.play(Write(ROW2))
        self.wait()
        
        #ROW3
        self.play(Write(ROW3))
        self.wait()
        
        #ROW4
        self.play(Write(ROW4))
        self.wait()
        
        #ROW5
        self.play(Write(ROW5))
        self.wait(3)
        
        self.play(FadeOut(*[subm for subm in self.get_mobject_family_members()]))
        
        Goodbye = MarkupText('Дякую за перегляд', gradient=(BLUE, YELLOW))
        
        self.play(Write(Goodbye))
        
        self.wait(1)
        self.play(FadeOut(Goodbye))
```
