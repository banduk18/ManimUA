```python
%%manim -qk -v WARNING G1_20

class G1_20(Scene):
    def construct(self):
        
        #Ttile and condition
        
        title = Text('§ 1. Чотирикутники - 20').scale(0.75)
        title.shift(UP*3.5)
        Condition = Text('У трикутнику ABC відомо, що ∠A = 44°, ∠B = 56°.\nБісектриса AK і BM трикутника перетинаються в точці O.\nЗнайдіть кути чотирикутника: 1)MOKC; 2) AOBC').scale(0.7)
        
        
        self.play(FadeIn(title))
        self.play(Write(Condition), run_time = 5)
        self.wait(2)
        self.play(FadeOut(Condition))
        
        #CONST
        
        prop = 0.4
        
        C = 10*prop
        B = 8.3*prop
        A = 5.74*prop
        
        #Dots_ABC
        
        dot_A = Dot()
    
        dot_B = dot_A.copy()
        dot_B.shift((UP*math.sin(44.01041871332267*DEGREES)+RIGHT*math.cos(44.01041871332267*DEGREES))*C)
        
        dot_C = dot_A.copy()
        dot_C.shift(RIGHT*B)
        dots = VGroup(*[dot_A, dot_B, dot_C])
        dots.shift(RIGHT*1.5)
        
        text_A = Text("A").scale(0.5).next_to(dot_A, DL)
        text_B = Text("B").scale(0.5).next_to(dot_B, UP)
        text_C = Text("C").scale(0.5).next_to(dot_C, DR)
        text_dots = VGroup(*[text_A, text_B, text_C])
               
        self.play(FadeIn(dots, text_dots))
        
        #Lines_ABC
        
        line_a = Line(dot_B, dot_C)
        line_b = Line(dot_A, dot_C)
        line_c = Line(dot_A, dot_B)
        lines = VGroup(*[line_a, line_b, line_c])
        
        self.play(Create(lines))
        
        #ABC Angles
        
        ABC_angle_A = Angle(line_b, line_c, radius = 0.4)
        ABC_angle_B = Angle(line_a, line_c, radius = 0.4, quadrant=(1,-1), other_angle=True)
        ABC_angle_C = Angle(line_b, line_a, radius = 0.4, quadrant=(-1,-1), other_angle=True)
        ABC_angles = VGroup(*[ABC_angle_A, ABC_angle_B, ABC_angle_C])
        
        self.play(Create(ABC_angles))
        
        ABC_value_A = Tex(r"44")
        ABC_value_A.scale(0.5).next_to(ABC_angle_A, RIGHT+UP*0.12)
        ABC_value_B = Tex(r"56")
        ABC_value_B.scale(0.5).next_to(ABC_angle_B, DOWN*0.30)
        ABC_values = VGroup(*[ABC_value_A, ABC_value_B])

        self.play(Write(ABC_values))
        
        ABC_value_C1 = Tex(r"180 - 44° - 56").scale(0.5).next_to(ABC_angle_C)
        ABC_value_C2 = Tex(r"180 - (44 + 56)").scale(0.5).next_to(ABC_angle_C)
        ABC_value_C3 = Tex(r"180 - 100").scale(0.5).next_to(ABC_angle_C)
        ABC_value_C4 = Tex(r"80").scale(0.5).next_to(ABC_angle_C)
        
        
        self.play(Write(ABC_value_C1))
        self.play(ReplacementTransform(ABC_value_C1,ABC_value_C2))
        self.wait(0.5)
        self.play(ReplacementTransform(ABC_value_C2,ABC_value_C3))
        self.wait(0.5)
        self.play(ReplacementTransform(ABC_value_C3, ABC_value_C4))
        self.wait(0.5)
        
        #RESULTS 
        #ΔABC
        ABC_result_title = Text("ΔABC")
        ABC_result_A = Text("∠A =")
        ABC_result_B = Text("∠B =")
        ABC_result_C = Text("∠C =")
        ABC_results = VGroup(*[ABC_result_title
                               , ABC_result_A, ABC_result_B
                               , ABC_result_C]).arrange(DOWN, buff = 0.5).scale(0.5)
        ABC_results.shift(LEFT*5, UP*2)
        
        self.play(FadeOut(ABC_angles),
                  Write(ABC_results)
                 )
        self.play(ABC_value_A.animate.next_to(ABC_result_A, RIGHT)
                 ,ABC_value_B.animate.next_to(ABC_result_B, RIGHT)
                 ,ABC_value_C4.animate.next_to(ABC_result_C, RIGHT))

        #Dot_K
        BK = C*A/(C+B)
        KC = A - BK
        
        dot_K = dot_C.copy()
        dot_K.shift(UP*math.sin(Angle(line_b, line_a, quadrant=(-1,1)).get_value()) + 
                     RIGHT*math.cos(Angle(line_b, line_a, quadrant=(-1,1)).get_value())*KC)
        line_K = Line(dot_A, dot_K)
        text_K = Text("K").scale(0.5).next_to(dot_K, RIGHT)
        
        self.play(Create(line_K))
        self.play(FadeIn(dot_K, text_K))
        self.wait()
        
        #A_bis
        
        BAK_angle_A = Angle(line_c, line_K, radius=0.4, quadrant=(1,1), other_angle=True)
        KAC_angle_A = Angle(line_K, line_b, radius=0.5, quadrant=(1,1), other_angle=True)
        A_bis = VGroup(*[BAK_angle_A,KAC_angle_A])
        self.play(Create(A_bis))
        self.wait()
        
        #Dot_M
        
        AM = C*B/(A+C)
        dot_M = dot_A.copy()
        dot_M.shift(RIGHT*AM)
        line_M = Line(dot_B, dot_M)
        text_M = Text("M").scale(0.5).next_to(dot_M, DOWN)
        
        self.play(Create(line_M))
        self.play(FadeIn(dot_M, text_M))
        
        #B_bis
        
        ABM_angle_B = Angle(line_c, line_M, radius = 0.4, quadrant=(-1,1))
        ABM_angle_B1 = Angle(line_c, line_M, radius = 0.5, quadrant=(-1,1))
        BMC_angle_B = Angle(line_M, line_a, radius = 0.5)
        BMC_angle_B1 = Angle(line_M, line_a, radius = 0.6)
        B_bis = VGroup(*[ABM_angle_B,ABM_angle_B1, BMC_angle_B,BMC_angle_B1])
        self.play(Create(B_bis))
        self.wait()
        
        #Dot_O
        AO = line_K.get_length() - (BK*line_K.get_length()/(C+AM))
        dot_O = dot_A.copy()
        dot_O.shift((UP*math.sin(ABC_angle_A.get_value()/2-0.08) +
                     RIGHT*math.cos(ABC_angle_A.get_value()/2))*AO)
        text_O = Text("O").scale(0.5).next_to(dot_O, UL)

        self.play(FadeIn(dot_O, text_O))
        
        #ΔACK
        #ΔACK_Title
        ACK_result_title = Text("ΔACK").set_color(YELLOW)
        ACK_result_A = Text("∠A =").set_color(YELLOW)
        ACK_result_C = Text("∠C =").set_color(YELLOW)
        ACK_result_K = Text("∠K =").set_color(YELLOW)
        ACK_results = VGroup(*[ACK_result_title
                              ,ACK_result_A
                              ,ACK_result_C
                              ,ACK_result_K]
                            ).arrange(DOWN, buff = 0.5).scale(0.5)
        ACK_results.shift(UP*2+LEFT*3)
        
        self.play(Write(ACK_results))
        
        #ΔACK_Angle
        ACK_angle_K = Angle(line_K, line_a, radius = 0.3, quadrant=(-1,1)).set_color(YELLOW)
        KAC_angle_A.set_color(YELLOW)
        ABC_angle_C.set_color(YELLOW)
        
        
        
        self.play(Create(ACK_angle_K)
                 ,KAC_angle_A.animate.set_color(YELLOW)
                 ,Create(ABC_angle_C)
                 )
        #ΔACK_Value
        ACK_value_A = Tex(r"22").scale(0.5).next_to(ACK_result_A,RIGHT).set_color(YELLOW)
        ACK_value_C = Tex(r"80").scale(0.5).next_to(ACK_result_C,RIGHT).set_color(YELLOW)
        ACK_value_K1 = Tex(r"180 - 22 - 80").scale(0.5).next_to(ACK_result_K, RIGHT).set_color(YELLOW)
        ACK_value_K2 = Tex(r"180 - (22 + 80)").scale(0.5).next_to(ACK_result_K, RIGHT).set_color(YELLOW)
        ACK_value_K3 = Tex(r"180 - 102").scale(0.5).next_to(ACK_result_K, RIGHT).set_color(YELLOW)
        ACK_value_K4 = Tex(r"78").scale(0.5).next_to(ACK_result_K, RIGHT).set_color(YELLOW)
        
        self.play(Write(ACK_value_A)
                 ,Write(ACK_value_C)
                 ,Write(ACK_value_K1)
                 )
        
        self.play(ReplacementTransform(ACK_value_K1, ACK_value_K2))
        self.wait(0.5)
        self.play(ReplacementTransform(ACK_value_K2, ACK_value_K3))
        self.wait(0.5)
        self.play(ReplacementTransform(ACK_value_K3, ACK_value_K4))
        self.wait()
        
        
        #ΔBMC
        #ΔBMC_Title
        BMC_result_title = Text("ΔBMC").set_color(RED)
        BMC_result_B = Text("∠B =").set_color(RED)
        BMC_result_M = Text("∠M =").set_color(RED)
        BMC_result_C = Text("∠C =").set_color(RED)
        BMC_results = VGroup(*[BMC_result_title
                              ,BMC_result_B
                              ,BMC_result_C
                              ,BMC_result_M]
                            ).arrange(DOWN, buff = 0.5).scale(0.5)
        BMC_results.shift(UP*2+LEFT)
        
        self.play(Write(BMC_results))
        
        #ΔBMC_Angle
        BMC_angle_M = Angle(line_b, line_M, radius= 0.2, quadrant=(1,-1)).set_color(RED)
        ABC_angle_C1 = Angle(line_b, line_a, radius = 0.2, quadrant=(-1,-1), other_angle=True).set_color(RED)
        
        self.play(Create(BMC_angle_M)
                 ,Create(ABC_angle_C1)
                 ,BMC_angle_B.animate.set_color(RED)
                 ,BMC_angle_B1.animate.set_color(RED)
                 )
        #ΔBMC_Value
        BMC_value_B = Tex(r"28").scale(0.5).next_to(BMC_result_B,RIGHT).set_color(RED)
        BMC_value_C = Tex(r"80").scale(0.5).next_to(BMC_result_C,RIGHT).set_color(RED)
        BMC_value_M1 = Tex(r"180 - 28 - 80").scale(0.5).next_to(BMC_result_M,RIGHT).set_color(RED)
        BMC_value_M2 = Tex(r"180 - (28 + 80)").scale(0.5).next_to(BMC_result_M,RIGHT).set_color(RED)
        BMC_value_M3 = Tex(r"180 - 108").scale(0.5).next_to(BMC_result_M,RIGHT).set_color(RED)
        BMC_value_M4 = Tex(r"72").scale(0.5).next_to(BMC_result_M,RIGHT).set_color(RED)
        
        self.play(Write(BMC_value_B)
                 ,Write(BMC_value_C)
                 ,Write(BMC_value_M1)
                 )
        
        self.play(ReplacementTransform(BMC_value_M1, BMC_value_M2))
        self.wait(0.5)
        self.play(ReplacementTransform(BMC_value_M2, BMC_value_M3))
        self.wait(0.5)
        self.play(ReplacementTransform(BMC_value_M3, BMC_value_M4))
        self.wait()
        
        #MOKC
        #MOKC_Title
        MOKC_result_title = Text("MOKC").set_color(ORANGE)
        MOKC_result_M = Text("∠M =").set_color(ORANGE)
        MOKC_result_O = Text("∠O =").set_color(ORANGE)
        MOKC_result_K = Text("∠K =").set_color(ORANGE)
        MOKC_result_C = Text("∠C =").set_color(ORANGE)
        MOKC_results = VGroup(*[MOKC_result_title
                               ,MOKC_result_M
                               ,MOKC_result_O
                               ,MOKC_result_K
                               ,MOKC_result_C]
                             ).arrange(DOWN, buff = 0.5).scale(0.5)
        MOKC_results.shift(DOWN+LEFT*5)
        
        self.play(Write(MOKC_results))
        
        #MOKC_Angle
        MOKC_angle_O = Angle(line_M, line_K, radius=0.2, quadrant=(1,1)).set_color(ORANGE)
        
        self.play(BMC_angle_B.animate.set_color(WHITE)
                 ,BMC_angle_B1.animate.set_color(WHITE)
                 ,KAC_angle_A.animate.set_color(WHITE)
                 ,ACK_angle_K.animate.set_color(ORANGE)
                 ,FadeOut(ABC_angle_C1)
                 ,BMC_angle_M.animate.set_color(ORANGE)
                 ,ABC_angle_C.animate.set_color(ORANGE)
                 ,Create(MOKC_angle_O)
                 )
        
        #MOKC_Value
        
        MOKC_value_M = Tex(r"72").scale(0.5).next_to(MOKC_result_M,RIGHT).set_color(ORANGE)
        MOKC_value_K = Tex(r"78").scale(0.5).next_to(MOKC_result_K,RIGHT).set_color(ORANGE)
        MOKC_value_C = Tex(r"80").scale(0.5).next_to(MOKC_result_C,RIGHT).set_color(ORANGE)
        MOKC_value_O1 = Tex(r"360 - 72 - 78 - 80").scale(0.5).next_to(MOKC_result_O,RIGHT).set_color(ORANGE)
        MOKC_value_O2 = Tex(r"360 - (72 + 78 + 80)").scale(0.5).next_to(MOKC_result_O,RIGHT).set_color(ORANGE)
        MOKC_value_O3 = Tex(r"360 - 230").scale(0.5).next_to(MOKC_result_O,RIGHT).set_color(ORANGE)
        MOKC_value_O4 = Tex(r"130").scale(0.5).next_to(MOKC_result_O,RIGHT).set_color(ORANGE)
        
        self.play(Write(MOKC_value_M)
                 ,Write(MOKC_value_O1)
                 ,Write(MOKC_value_K)
                 ,Write(MOKC_value_C)
                 )
        
        self.play(ReplacementTransform(MOKC_value_O1, MOKC_value_O2))
        self.wait(0.5)
        self.play(ReplacementTransform(MOKC_value_O2, MOKC_value_O3))
        self.wait(0.5)
        self.play(ReplacementTransform(MOKC_value_O3, MOKC_value_O4))
        self.wait()
        
        #AOBC
        #AOBC_Title
        
        AOBC_result_title = Text("AOBC").set_color(PURPLE)
        AOBC_result_A = Text("∠A =").set_color(PURPLE)
        AOBC_result_O = Text("∠O =").set_color(PURPLE)
        AOBC_result_B = Text("∠B =").set_color(PURPLE)
        AOBC_result_C = Text("∠C =").set_color(PURPLE)
        AOBC_results = VGroup(*[AOBC_result_title
                               ,AOBC_result_A
                               ,AOBC_result_O
                               ,AOBC_result_B
                               ,AOBC_result_C]
                             ).arrange(DOWN, buff = 0.5).scale(0.5)
        AOBC_results.shift(DOWN+LEFT*3)
        
        self.play(Write(AOBC_results))
        
        #AOBC_Angle
        MOKC_angle_O1 = Angle(line_M, line_K, radius=0.2, quadrant=(-1,-1), other_angle=True).set_color(PURPLE)
        
        self.play(FadeOut(ACK_angle_K)
                 ,FadeOut(BMC_angle_M)
                 ,ABC_angle_C.animate.set_color(PURPLE)
                 ,BMC_angle_B.animate.set_color(PURPLE)
                 ,BMC_angle_B1.animate.set_color(PURPLE)
                 ,KAC_angle_A.animate.set_color(PURPLE)
                 ,ReplacementTransform(MOKC_angle_O, MOKC_angle_O1)
                 )
        #AOBC_Value
        AOBC_value_A = Tex(r"22").scale(0.5).next_to(AOBC_result_A,RIGHT).set_color(PURPLE)
        AOBC_value_B = Tex(r"28").scale(0.5).next_to(AOBC_result_B,RIGHT).set_color(PURPLE)
        AOBC_value_C = Tex(r"80").scale(0.5).next_to(AOBC_result_C,RIGHT).set_color(PURPLE)
        AOBC_value_O1 = Tex(r"360 - 22 - 28 - 80").scale(0.5).next_to(AOBC_result_O,RIGHT).set_color(PURPLE)
        AOBC_value_O2 = Tex(r"360 - (22 + 28 + 80)").scale(0.5).next_to(AOBC_result_O,RIGHT).set_color(PURPLE)
        AOBC_value_O3 = Tex(r"360 - 130").scale(0.5).next_to(AOBC_result_O,RIGHT).set_color(PURPLE)
        AOBC_value_O4 = Tex(r"230").scale(0.5).next_to(AOBC_result_O,RIGHT).set_color(PURPLE)
        
        self.play(Write(AOBC_value_A)
                 ,Write(AOBC_value_O1)
                 ,Write(AOBC_value_B)
                 ,Write(AOBC_value_C)
                 )
        
        self.play(ReplacementTransform(AOBC_value_O1, AOBC_value_O2))
        self.wait(0.5)
        self.play(ReplacementTransform(AOBC_value_O2, AOBC_value_O3))
        self.wait(0.5)
        self.play(ReplacementTransform(AOBC_value_O3, AOBC_value_O4))
        
        #FINISH
        
        FINISH_result_title = Text("Відповідь:")
        FINISH_result_MOKC = Text("1) MOKC: 72°, 130°, 78°, 80°")
        FINISH_result_AOBC = Text("2) AOBC: 22°, 230°, 28°, 80°")
        FINISH_results = VGroup(*[
            FINISH_result_title,
            FINISH_result_MOKC,
            FINISH_result_AOBC
        ]).arrange(DOWN, buff = 0.5).scale(0.5)
        FINISH_results.shift(2*DOWN+RIGHT)
        
        self.play(Write(FINISH_results))
        self.wait(2)
        
        self.clear()
        Goodbye = MarkupText('Дякую за перегляд', gradient=(BLUE, YELLOW))
        
        self.play(Write(Goodbye))
        
        self.wait(2)
        self.play(FadeOut(Goodbye))
```

```python
%%manim -qh -v WARNING G1_21

class G1_21(Scene):
    def construct(self):
        
        #Title
        
        title = Text('§ 1. Чотирикутники - 21').scale(0.75)
        title.shift(UP*3.5)
        Condition = Text('У трикутнику ABC відомо, що ∠A = 36°, ∠B = 72°.\nВисоти AE і BF трикутника перетинаються в точці H.\nЗнайдіть кути чотирикутника: 1)CFHE; 2) ACBH').scale(0.7)
        
        self.play(FadeIn(title))
        self.play(Write(Condition), run_time = 5)
        self.wait(2)
        self.play(FadeOut(Condition))
        
        #CONST
        
        prop = 4
        A = 0.618033983*prop
        B = 1*prop
        C = 1*prop
        
        #ΔABC
        
        dot_A = Dot()
    
        dot_B = dot_A.copy()
        dot_B.shift((UP*math.sin(36*DEGREES)+RIGHT*math.cos(36*DEGREES))*C)
        
        dot_C = dot_A.copy()
        dot_C.shift(RIGHT*B)
        dots = VGroup(*[dot_A, dot_B, dot_C])
        dots.shift(RIGHT*0.5)
        
        text_A = Text("A").scale(0.5).next_to(dot_A, DL)
        text_B = Text("B").scale(0.5).next_to(dot_B, UP)
        text_C = Text("C").scale(0.5).next_to(dot_C, DR)
        text_dots = VGroup(*[text_A, text_B, text_C])
               
        self.play(FadeIn(dots, text_dots))
        
        #Lines
        line_a = Line(dot_B, dot_C)
        line_b = Line(dot_A, dot_C)
        line_c = Line(dot_A, dot_B)
        lines = VGroup(*[line_a, line_b, line_c])
        
        self.play(Create(lines))
        
        #ABC Angles
        
        ABC_angle_A = Angle(line_b, line_c, radius = 0.2)
        ABC_angle_B = Angle(line_a, line_c, radius = 0.2, quadrant=(1,-1), other_angle=True)
        ABC_angle_C = Angle(line_b, line_a, radius = 0.2, quadrant=(-1,-1), other_angle=True)
        ABC_angles = VGroup(*[ABC_angle_A, ABC_angle_B, ABC_angle_C])
        
        self.play(Create(ABC_angles))
        
        #ABC Values
        
        ABC_value_A = Text("36")
        ABC_value_A.scale(0.5).next_to(ABC_angle_A, RIGHT+UP*0.01)
        ABC_value_B = Text("72")
        ABC_value_B.scale(0.5).next_to(ABC_angle_B, DOWN*0.30)
        ABC_values = VGroup(*[ABC_value_A, ABC_value_B])

        self.play(Write(ABC_values))
        
        ABC_value_C1 = Text("180 - 36 - 72").scale(0.5).next_to(ABC_angle_C)
        ABC_value_C2 = Text("180 - (36 + 72)").scale(0.5).next_to(ABC_angle_C)
        ABC_value_C3 = Text("180 - 108").scale(0.5).next_to(ABC_angle_C)
        ABC_value_C4 = Text("72").scale(0.5).next_to(ABC_angle_C)
        
        
        self.play(Write(ABC_value_C1))
        self.play(ReplacementTransform(ABC_value_C1,ABC_value_C2))
        self.wait(0.5)
        self.play(ReplacementTransform(ABC_value_C2,ABC_value_C3))
        self.wait(0.5)
        self.play(ReplacementTransform(ABC_value_C3, ABC_value_C4))
        
        self.wait()
        
        #RESULTS 
        #ΔABC
        ABC_result_title = Text("ΔABC")
        ABC_result_A = Text("∠A =")
        ABC_result_B = Text("∠B =")
        ABC_result_C = Text("∠C =")
        ABC_results = VGroup(*[ABC_result_title
                               , ABC_result_A, ABC_result_B
                               , ABC_result_C]).arrange(DOWN, buff = 0.5).scale(0.5)
        ABC_results.shift(LEFT*5, UP*2)
        
        
        
        self.play(
                  Write(ABC_results)
                 )
        self.play(ABC_value_A.animate.next_to(ABC_result_A, RIGHT)
                 ,ABC_value_B.animate.next_to(ABC_result_B, RIGHT)
                 ,ABC_value_C4.animate.next_to(ABC_result_C, RIGHT))
        
        triangle = VGroup(*[
            dots,
            lines,
            ABC_angles,
        ])
        
        self.play(triangle.animate.rotate(-109*DEGREES).shift(RIGHT*3))
        self.play(text_A.animate.next_to(dot_A)
                 ,text_B.animate.next_to(dot_B)
                 ,text_C.animate.next_to(dot_C,LEFT)
                 )
        self.play(Circumscribe(ABC_value_C4)
                 ,Circumscribe(ABC_value_B)
                 ,Flash(ABC_angle_B)
                 ,Flash(ABC_angle_C)
                 )
        self.wait()
        
        #Isosceles
        
        isos_lane1 = Line(start = [-0.1, 0, 0], end = [0.1, 0,0])
        isos_lane2 = isos_lane1.copy()
        isos_lane1.rotate(-18*DEGREES)
        isos_lane2.rotate(18*DEGREES)
        
        isos_lane1.shift(line_b.get_midpoint())
        isos_lane2.shift(line_c.get_midpoint())
        
        self.play(Create(isos_lane1),Create(isos_lane2))
        
        self.wait()
                
        #Dot E
        
        dot_E = dot_C.copy()
        dot_E.shift(RIGHT*(line_a.get_length()/2+0.05)+DOWN*0.02)
        text_E = Text("E").scale(0.5).next_to(dot_E, DOWN)
        self.play(FadeIn(dot_E)
                 ,FadeOut(ABC_angle_A)
                 ,FadeOut(ABC_angle_B)
                 ,FadeOut(ABC_angle_C)
                 ,FadeIn(text_E)
                 )
        
        line_E = Line(dot_A, dot_E)
        angle_E = RightAngle(line_a, line_E, length=0.2, quadrant = (-1,-1))
        self.play(Create(line_E)
                 ,Create(angle_E)
                 )
        
        self.wait()
        
        #Dot F
        
        dot_F = dot_B.copy()
        dot_F.shift((LEFT*(math.cos(PI/2 + ABC_angle_C.get_value()))
                   +UP*(math.sin(PI/2 + ABC_angle_C.get_value())))*line_a.get_length())
        text_F = Text("F").scale(0.5).next_to(dot_F, LEFT)
        self.play(FadeIn(dot_F)
                 ,FadeIn(text_F)
                 )
        
        line_F = Line(dot_B, dot_F)
        angle_F = RightAngle(line_F, line_b, length = 0.2, quadrant=(-1,-1))
        self.play(Create(line_F)
                 ,Create(angle_F)
                 )
        
        self.wait()
        
        #Dot H
        
        dot_H = dot_B.copy()
        dot_H.shift(((LEFT
                   +UP*(math.sin(PI/2 + ABC_angle_C.get_value())))*line_a.get_length()/1.9))

        text_H = Text("H").scale(0.5).next_to(dot_H,UR)
        self.play(FadeIn(dot_H)
                 ,FadeIn(text_H)
                 )
        
        #CFHE
        
        #CFHE_Title
        CFHE_result_title = Text("CFHE").set_color(ORANGE)
        CFHE_result_C = Text("∠C =").set_color(ORANGE)
        CFHE_result_F = Text("∠F =").set_color(ORANGE)
        CFHE_result_H = Text("∠H =").set_color(ORANGE)
        CFHE_result_E = Text("∠E =").set_color(ORANGE)
        CFHE_results = VGroup(*[CFHE_result_title
                               ,CFHE_result_C
                               ,CFHE_result_F
                               ,CFHE_result_H
                               ,CFHE_result_E]
                             ).arrange(DOWN, buff = 0.5).scale(0.5)
        CFHE_results.shift(DOWN+LEFT*5)
        
        self.play(Write(CFHE_results))
        self.wait()
        
        #CFHE_Angles
        
        CFHE_angle_C = ABC_angle_C.copy()
        CFHE_angle_C.set_color(ORANGE)
        CFHE_angle_F = RightAngle(line_F, line_b, length = 0.2, quadrant=(-1,1)).set_color(ORANGE)
        CFHE_angle_E = RightAngle(line_a, line_E, length=0.15, quadrant = (1,-1)).set_color(ORANGE)
        CFHE_angle_H = Angle(line_F, line_E, radius=0.1, quadrant = (1,1)).set_color(ORANGE)
        
        self.play(Create(CFHE_angle_C)
                 ,Create(CFHE_angle_F)
                 ,Create(CFHE_angle_E)
                 ,Create(CFHE_angle_H)
                 )
        self.wait()
        
        #CFHE_Indicate
        
        CF = Line(dot_C, dot_F) 
        FH = Line(dot_F, dot_H)
        HE = Line(dot_H, dot_E)
        EC = Line(dot_E, dot_C)
        
        CFHE = VGroup(*[
            CF, FH, HE, EC,
            dot_C, dot_F, dot_H, dot_E,
            text_C, text_F, text_H, text_E
        ])
        
        self.play(Indicate(CFHE, color = ORANGE))
        self.play(FadeOut(CF, FH, HE, EC))
        self.wait()
        
        #CFHE_Values
        
        CFHE_value_C = Tex(r"72").scale(0.5).next_to(CFHE_result_C,RIGHT).set_color(ORANGE)
        CFHE_value_F = Tex(r"90").scale(0.5).next_to(CFHE_result_F,RIGHT).set_color(ORANGE)
        CFHE_value_E = Tex(r"90").scale(0.5).next_to(CFHE_result_E,RIGHT).set_color(ORANGE)
        CFHE_value_H1 = Tex(r"360 - 90 - 90 - 72").scale(0.5).next_to(CFHE_result_H,RIGHT).set_color(ORANGE)
        CFHE_value_H2 = Tex(r"360 - (90 + 90 + 72)").scale(0.5).next_to(CFHE_result_H,RIGHT).set_color(ORANGE)
        CFHE_value_H3 = Tex(r"360 - 252").scale(0.5).next_to(CFHE_result_H,RIGHT).set_color(ORANGE)
        CFHE_value_H4 = Tex(r"108").scale(0.5).next_to(CFHE_result_H,RIGHT).set_color(ORANGE)
        
        self.play(Write(CFHE_value_C)
                 ,Write(CFHE_value_H1)
                 ,Write(CFHE_value_F)
                 ,Write(CFHE_value_E)
                 )
        
        self.wait()
        self.play(ReplacementTransform(CFHE_value_H1, CFHE_value_H2))
        self.wait(0.5)
        self.play(ReplacementTransform(CFHE_value_H2, CFHE_value_H3))
        self.wait(0.5)
        self.play(ReplacementTransform(CFHE_value_H3, CFHE_value_H4))
        self.wait()

        #ΔAEC
        
        #ΔAEC_Title
        AEC_result_title = Text("ΔAEC").set_color(RED)
        AEC_result_A = Text("∠A =").set_color(RED)
        AEC_result_E = Text("∠E =").set_color(RED)
        AEC_result_C = Text("∠C =").set_color(RED)
        AEC_results = VGroup(*[AEC_result_title
                              ,AEC_result_A
                              ,AEC_result_E
                              ,AEC_result_C]
                            ).arrange(DOWN, buff = 0.5).scale(0.5)
        AEC_results.shift(UP*2+LEFT*3)
        
        self.play(Write(AEC_results))
        self.wait()
        
        #AEC_Angles
        AEC_angle_A = Angle(line_b, line_E, radius = 0.6).set_color(RED)
        AEC_angle_C = CFHE_angle_C
        AEC_angle_E = CFHE_angle_E
        
        
        
        self.play(FadeOut(CFHE_angle_F)
                 ,FadeOut(CFHE_angle_H)
                 ,Create(AEC_angle_A)
                 ,AEC_angle_E.animate.set_color(RED)
                 ,AEC_angle_C.animate.set_color(RED)
                 )
        self.wait()
        
        #AEC_Indicate
        
        CA = Line(dot_A, dot_C) 
        AE = Line(dot_A, dot_E)
        EC = Line(dot_E, dot_C)
        
        AEC = VGroup(*[
            CA, AE, EC, #sides
            dot_C, dot_A, dot_E, #dots
            text_C, text_A, text_E #texts
        ])
        
        self.play(Indicate(AEC, color=RED))
        self.play(FadeOut(CA, AE, EC))
        self.wait()
        
        #AEC_Values
        
        AEC_value_C = Tex(r"72").scale(0.5).next_to(AEC_result_C,RIGHT).set_color(RED)
        AEC_value_E = Tex(r"90").scale(0.5).next_to(AEC_result_E,RIGHT).set_color(RED)
        AEC_value_A1 = Tex(r"180 - 90 - 72").scale(0.5).next_to(AEC_result_A,RIGHT).set_color(RED)
        AEC_value_A2 = Tex(r"180 - (90 + 72)").scale(0.5).next_to(AEC_result_A,RIGHT).set_color(RED)
        AEC_value_A3 = Tex(r"180 - 162").scale(0.5).next_to(AEC_result_A,RIGHT).set_color(RED)
        AEC_value_A4 = Tex(r"18").scale(0.5).next_to(AEC_result_A,RIGHT).set_color(RED)
        
        self.play(Write(AEC_value_C)
                 ,Write(AEC_value_E)
                 ,Write(AEC_value_A1)
                 )
        
        self.wait()
        self.play(ReplacementTransform(AEC_value_A1, AEC_value_A2))
        self.wait(0.5)
        self.play(ReplacementTransform(AEC_value_A2, AEC_value_A3))
        self.wait(0.5)
        self.play(ReplacementTransform(AEC_value_A3, AEC_value_A4))
        self.wait()
        
        #ΔBFC
        #BFC_Title
        
        BFC_result_title = Text("ΔBFC").set_color(YELLOW)
        BFC_result_B = Text("∠B =").set_color(YELLOW)
        BFC_result_F = Text("∠F =").set_color(YELLOW)
        BFC_result_C = Text("∠C =").set_color(YELLOW)
        BFC_results = VGroup(*[BFC_result_title
                              ,BFC_result_B
                              ,BFC_result_F
                              ,BFC_result_C]
                            ).arrange(DOWN, buff = 0.5).scale(0.5)
        BFC_results.shift(UP*2+LEFT*1)
        
        self.play(Write(BFC_results))
        self.wait()
        
        #BFC_Angles
        BFC_angle_B = Angle(line_a, line_F, radius = 0.6, other_angle=True).set_color(YELLOW)
        BFC_angle_F = CFHE_angle_F
        BFC_angle_C = CFHE_angle_C
        
        BFC_angle_F.set_color(YELLOW)
        
        self.play(FadeIn(BFC_angle_F)
                 ,FadeOut(AEC_angle_E)
                 ,FadeOut(AEC_angle_A)
                 ,BFC_angle_C.animate.set_color(YELLOW)
                 ,Create(BFC_angle_B)
                 )
        self.wait()
        
        #BFC_Indicate
        
        CF = Line(dot_C, dot_F) 
        FB = Line(dot_F, dot_B)
        BC = Line(dot_B, dot_C)
        
        BFC = VGroup(*[
            CF, FB, BC,
            dot_B, dot_F, dot_C,
            text_B, text_F, text_C,
        ])
        
        self.play(Indicate(BFC, color = YELLOW))
        self.play(FadeOut(CF, FB, BC))
        self.wait()
        
        #BFC_Values
        
        BFC_value_C = Tex(r"72").scale(0.5).next_to(BFC_result_C,RIGHT).set_color(YELLOW)
        BFC_value_F = Tex(r"90").scale(0.5).next_to(BFC_result_F,RIGHT).set_color(YELLOW)
        BFC_value_B1 = Tex(r"180 - 90 - 72").scale(0.5).next_to(BFC_result_B,RIGHT).set_color(YELLOW)
        BFC_value_B2 = Tex(r"180 - (90 + 72)").scale(0.5).next_to(BFC_result_B,RIGHT).set_color(YELLOW)
        BFC_value_B3 = Tex(r"180 - 162").scale(0.5).next_to(BFC_result_B,RIGHT).set_color(YELLOW)
        BFC_value_B4 = Tex(r"18").scale(0.5).next_to(BFC_result_B,RIGHT).set_color(YELLOW)
        
        self.play(Write(BFC_value_C)
                 ,Write(BFC_value_F)
                 ,Write(BFC_value_B1)
                 )
        
        self.wait()
        self.play(ReplacementTransform(BFC_value_B1, BFC_value_B2))
        self.wait(0.5)
        self.play(ReplacementTransform(BFC_value_B2, BFC_value_B3))
        self.wait(0.5)
        self.play(ReplacementTransform(BFC_value_B3, BFC_value_B4))
        self.wait()
        
        #ACBH

        #ACBH Title
        ACBH_result_title = Text("ACBH").set_color(PURPLE)
        ACBH_result_A = Text("∠A =").set_color(PURPLE)
        ACBH_result_C = Text("∠C =").set_color(PURPLE)
        ACBH_result_B = Text("∠B =").set_color(PURPLE)
        ACBH_result_H = Text("∠H =").set_color(PURPLE)
        ACBH_results = VGroup(*[ACBH_result_title
                               ,ACBH_result_A
                               ,ACBH_result_C
                               ,ACBH_result_B
                               ,ACBH_result_H]
                             ).arrange(DOWN, buff = 0.5).scale(0.5)
        ACBH_results.shift(DOWN+LEFT*3)
        
        self.play(Write(ACBH_results))
        self.wait()
        
        #ACBH_Angles
        
        ACBH_angle_A = AEC_angle_A.copy()
        ACBH_angle_A.set_color(PURPLE)
        ACBH_angle_C = ABC_angle_C.copy()
        ACBH_angle_B = BFC_angle_B.copy()
        ACBH_angle_H = Angle(line_F, line_E, radius=0.2, quadrant = (-1,-1), other_angle=True).set_color(PURPLE)
        
        self.play(ACBH_angle_C.animate.set_color(PURPLE)
                 ,ACBH_angle_B.animate.set_color(PURPLE)
                 ,Create(ACBH_angle_A)
                 ,Create(ACBH_angle_H)
                 ,FadeOut(BFC_angle_F)
                 )
        self.wait()
        
        #ACBH_Indicate
        
        AC = Line(dot_A, dot_C) 
        CB = Line(dot_C, dot_B)
        BH = Line(dot_B, dot_H)
        HA = Line(dot_H, dot_A)
        
        ACBH = VGroup(*[
            AC, CB, BH, HA,
            dot_A, dot_C, dot_B, dot_H,
            text_A, text_C, text_B, text_H
        ])
        
        self.play(Indicate(ACBH, color = PURPLE))
        self.play(FadeOut(AC, CB, BH, HA))
        self.wait()
        
        #ACBH Values
        
        ACBH_value_A = Tex(r"18").scale(0.5).next_to(ACBH_result_A,RIGHT).set_color(PURPLE)
        ACBH_value_C = Tex(r"72").scale(0.5).next_to(ACBH_result_C,RIGHT).set_color(PURPLE)
        ACBH_value_B = Tex(r"18").scale(0.5).next_to(ACBH_result_B,RIGHT).set_color(PURPLE)
        ACBH_value_H1 = Tex(r"360 - 18 - 72 - 18").scale(0.5).next_to(ACBH_result_H,RIGHT).set_color(PURPLE)
        ACBH_value_H2 = Tex(r"360 - (18 + 72 + 18)").scale(0.5).next_to(ACBH_result_H,RIGHT).set_color(PURPLE)
        ACBH_value_H3 = Tex(r"360 - 108").scale(0.5).next_to(ACBH_result_H,RIGHT).set_color(PURPLE)
        ACBH_value_H4 = Tex(r"252").scale(0.5).next_to(ACBH_result_H,RIGHT).set_color(PURPLE)
        
        self.play(Write(ACBH_value_A)
                 ,Write(ACBH_value_H1)
                 ,Write(ACBH_value_C)
                 ,Write(ACBH_value_B)
                 )
        
        self.wait()
        self.play(ReplacementTransform(ACBH_value_H1, ACBH_value_H2))
        self.wait(0.5)
        self.play(ReplacementTransform(ACBH_value_H2, ACBH_value_H3))
        self.wait(0.5)
        self.play(ReplacementTransform(ACBH_value_H3, ACBH_value_H4))
        self.wait()
        
        FINISH_result_title = Text("Відповідь:")
        FINISH_result_CFHE = Text("1) CFHE: 72, 130, 78, 80")
        FINISH_result_ACBH = Text("2) ACBH: 22, 230, 28, 80")
        FINISH_results = VGroup(*[
            FINISH_result_title,
            FINISH_result_CFHE,
            FINISH_result_ACBH
        ]).arrange(DOWN, buff = 0.5).scale(0.5)
        FINISH_results.shift(DOWN+RIGHT)
        
        self.play(Write(FINISH_results))
        self.wait(2)
        
        self.clear()
        Goodbye = MarkupText('Дякую за перегляд', gradient=(BLUE, YELLOW))
        
        self.play(Write(Goodbye))
        
        self.wait(1)
        self.play(FadeOut(Goodbye))
```

```python
%%manim -qh -v WARNING G1_22

class G1_22(Scene):
    def construct(self):
        
        #Title
        
        title = Text('§ 1. Чотирикутники - 22').scale(0.75)
        self.add_foreground_mobject(title)
        title.shift(UP*3.5)
        Condition = Text('Знайдіть діагональ Чотирикутника, якщо його периметр дорівнює 80 см,\nа перемитри трикутників, на які ця діагональ розбиває даний чотирикутник,\nдорівнюють 36 см і 64 см').scale(0.5)
        self.play(FadeIn(title))
        self.play(Write(Condition), run_time = 5)
        self.wait(2)
        self.play(FadeOut(Condition))
        
        #DOTS
        dot_A = Dot()
        
        dot_B = dot_A.copy()
        dot_B.shift((UP*math.sin(PI/3)+RIGHT*math.cos(PI/3))*3)
        
        dot_C = dot_B.copy()
        dot_C.shift((UP*math.sin(PI/-15)+RIGHT*math.cos(PI/-15))*4)
        
        dot_D = dot_A.copy()
        dot_D.shift(RIGHT*6)
        
        dots = VGroup(*[
            dot_A, dot_B, dot_C, dot_D
        ])
        
        self.play(Create(dots))
        #Lines
        
        AB = Line(dot_A, dot_B)
        BC = Line(dot_B, dot_C)
        CD = Line(dot_C, dot_D)
        DA = Line(dot_D, dot_A)
        BD = Line(dot_B, dot_D)
        
        lines = VGroup(*[
            AB, BC, CD, DA, BD
        ])
        
        self.play(Create(lines))
        #Text
        
        text_A = Tex("A").scale(0.75).next_to(dot_A,DL)
        text_B = Tex("B").scale(0.75).next_to(dot_B,UL)
        text_C = Tex("C").scale(0.75).next_to(dot_C,UR)
        text_D = Tex("D").scale(0.75).next_to(dot_D,DR)

        text_dots = VGroup(*[
            text_A, text_B, text_C, text_D
        ])
        
        self.play(Write(text_dots))
        self.wait()
        
        P_abd = Tex("$P_{ABD} = $").scale(0.75).set_color(YELLOW)
        P_bcd = Tex("$P_{BCD} = $").scale(0.75).set_color(RED)
        P_abcd = Tex("$P_{ABCD} = $").scale(0.75).set_color(ORANGE)
        
        perimeters = VGroup(*[
            P_abcd, P_abd, P_bcd
        ]).arrange(DOWN, buff = 0.5)
        
        perimeters.shift(UP*2+LEFT*5)
        
        self.play(Write(perimeters))
        
        ABCD = VGroup(*[
            AB, BC, CD, DA,
            dot_A, dot_B, dot_C, dot_D,
            text_A, text_B, text_C, text_D,
        ])
        
        ABCD_text_AB = Tex("$AB+$").scale(0.75).set_color(ORANGE)
        ABCD_text_BC = Tex("$BC+$").scale(0.75).set_color(ORANGE)
        ABCD_text_CD = Tex("$CD+$").scale(0.75).set_color(ORANGE)
        ABCD_text_DA = Tex("$DA$").scale(0.75).set_color(ORANGE)
        text_per_ABCD = Tex("$=80$").scale(0.75).set_color(ORANGE)
        
        text_ABCD = VGroup(*[
            ABCD_text_AB, ABCD_text_BC, ABCD_text_CD, ABCD_text_DA, text_per_ABCD
        ]).arrange(RIGHT, buff=0.1).next_to(P_abcd, RIGHT)
        
        AB1 = AB.copy()
        BC1 = BC.copy()
        CD1 = CD.copy()
        DA1 = DA.copy()
        
        self.play(Indicate(ABCD, color = ORANGE))
        self.play(AB1.animate.move_to(ABCD_text_AB),ReplacementTransform(AB1,ABCD_text_AB)
                 ,BC1.animate.move_to(ABCD_text_BC),ReplacementTransform(BC1,ABCD_text_BC)
                 ,CD1.animate.move_to(ABCD_text_CD),ReplacementTransform(CD1,ABCD_text_CD)
                 ,DA1.animate.move_to(ABCD_text_DA),ReplacementTransform(DA1,ABCD_text_DA)
                 )
        self.play(Write(text_per_ABCD))
        self.wait()
        #Indicate_ABD
        
        ABD = VGroup(*[
            AB, BD, DA,
            dot_A, dot_B, dot_D,
            text_A, text_B, text_D
        ])
        
        ABD_text_AB = Tex("$AB+$").scale(0.75).set_color(YELLOW)
        ABD_text_BD = Tex("$BD+$").scale(0.75).set_color(YELLOW)
        ABD_text_DA = Tex("$DA$").scale(0.75).set_color(YELLOW)
        text_per_ABD = Tex("$=64$").scale(0.75).set_color(YELLOW)
        
        
        AB2 = AB.copy()
        BD1 = BD.copy()
        DA2 = DA.copy()
        
        text_ABD = VGroup(*[
            ABD_text_AB, ABD_text_BD, ABD_text_DA, text_per_ABD
        ]).arrange(RIGHT, buff = 0.1).next_to(P_abd, RIGHT)
        
        self.play(Indicate(ABD, color = YELLOW))
        self.play(AB2.animate.move_to(ABD_text_AB),ReplacementTransform(AB2,ABD_text_AB)
                 ,BD1.animate.move_to(ABD_text_BD),ReplacementTransform(BD1,ABD_text_BD)
                 ,DA2.animate.move_to(ABD_text_DA),ReplacementTransform(DA2,ABD_text_DA)
                 )
        self.play(Write(text_per_ABD))
        
        self.wait()
        
        #Indicate_BCD
        
        BCD = VGroup(*[
            BC, CD, BD,
            dot_B, dot_C, dot_D,
            text_B, text_C, text_D
        ])
        
        BCD_text_BC = Tex("$BC+$").scale(0.75).set_color(RED)
        BCD_text_CD = Tex("$CD+$").scale(0.75).set_color(RED)
        BCD_text_BD = Tex("$BD$").scale(0.75).set_color(RED)
        text_per_BCD = Tex("$=36$").scale(0.75).set_color(RED)
        
        BC2 = BC.copy()
        BD2 = BD.copy()
        CD2 = CD.copy()
        
        text_BCD = VGroup(*[
            BCD_text_BC, BCD_text_CD, BCD_text_BD, text_per_BCD
        ]).arrange(RIGHT, buff = 0.1).next_to(P_bcd, RIGHT)
        
        self.play(Indicate(BCD, color = RED))
        self.play(BC2.animate.move_to(BCD_text_BC),ReplacementTransform(BC2,BCD_text_BC)
                 ,BD2.animate.move_to(BCD_text_BD),ReplacementTransform(BD2,BCD_text_BD)
                 ,CD2.animate.move_to(BCD_text_CD),ReplacementTransform(CD2,BCD_text_CD)
                 )
        self.play(Write(text_per_BCD))
                
        #Row_1
        TEXT1 = Tex("$P_{ABD} + P_{BCD}$").scale(0.75)
        TEXT1.shift(LEFT*5+DOWN)
        
        TEXT2 = Tex("$(AB + BD + DA) + (BC + CD + BD)$").scale(0.75)
        TEXT_eq_1 = Tex("$=$").scale(0.75)

        self.play(Write(TEXT1))
        self.play(Write(TEXT_eq_1.next_to(TEXT1,RIGHT)))
        self.play(Write(TEXT2.next_to(TEXT_eq_1,RIGHT)))
        self.wait()
        
        #Row_2
        TEXT3 = Tex("$64+36$").scale(0.75)
        TEXT3.shift(LEFT*5.5+DOWN*1.75)
        
        TEXT4 = Tex("$AB + BC + CD + DA$").scale(0.75)
        TEXT5 = Tex("$+ 2BD$").scale(0.75)
        TEXT_eq_2 = Tex("$=$").scale(0.75)
    
        self.play(Write(TEXT3))
        self.play(Write(TEXT_eq_2.next_to(TEXT3,RIGHT)))
        self.play(Write(TEXT4.next_to(TEXT_eq_2,RIGHT)),
                  Write(TEXT5.next_to(TEXT4,RIGHT*0.5)))
        
        self.wait()
        #Row_2.1
        TEXT6 = Tex("$100$").scale(0.75)
        TEXT6.shift(LEFT*5.5+DOWN*1.75)
        TEXT7 = Tex("$80$").scale(0.75)
        TEXT7.next_to(TEXT_eq_2,RIGHT)
        TEXT8 = Tex("$2BD$").scale(0.75)
        TEXT8.next_to(TEXT_eq_2,RIGHT)
        TEXT10 = Tex("$20$").scale(0.75)
        TEXT10.next_to(TEXT_eq_2,LEFT)
        
        TEXT11 = Tex("$10$").scale(0.75)
        TEXT11.next_to(TEXT_eq_2,LEFT)
        
        TEXT12 = Tex("$- 80$").scale(0.75)
        TEXT12.shift(LEFT*5.25+DOWN*1.75)
        
        TEXT13 = Tex("$BD$").scale(0.75)
        TEXT13.next_to(TEXT_eq_2,RIGHT)
        
        
        self.play(Circumscribe(TEXT4)
                 ,Circumscribe(VGroup(*[
                     ABCD_text_AB, ABCD_text_BC, ABCD_text_CD, ABCD_text_DA
                 ])))
        self.wait()
        self.play(ReplacementTransform(TEXT3,TEXT6)
                 ,ReplacementTransform(TEXT4,TEXT7)
                 ,TEXT5.animate.next_to(TEXT7,RIGHT)
                 )
        self.wait()
        self.play(TEXT6.animate.shift(LEFT*0.5)
                 ,Transform(TEXT7, TEXT12, path_arc= 90*DEGREES)
                 ,Transform(TEXT5, TEXT8))
        
        self.wait()
        self.play(ReplacementTransform(TEXT6,TEXT10),
                 FadeOut(TEXT7))
        
        self.wait()
        
        self.play(ReplacementTransform(TEXT5, TEXT13)
                 ,ReplacementTransform(TEXT10, TEXT11)
                 )
        
        self.wait()
        
        #FINISH
        
        FINISH_result_title = Text("Відповідь:")
        FINISH_result_BD = Text("10")
        FINISH_results = VGroup(*[
            FINISH_result_title,
            FINISH_result_BD
        ]).arrange(RIGHT, buff = 0.5).scale(0.5)
        FINISH_results.shift(2*DOWN+RIGHT)
        
        self.play(Write(FINISH_results))
        self.wait(2)
        
        self.clear()
        Goodbye = MarkupText('Дякую за перегляд', gradient=(BLUE, YELLOW))
        
        self.play(Write(Goodbye))
        
        self.wait(1)
        self.play(FadeOut(Goodbye))
        
```

```python
%%manim -qh -v WARNING G1_24_final

class G1_24_final(Scene):
    def construct(self):
        
        #Title
        
        title = Text('§ 1. Чотирикутники - 24').scale(0.75)
        title.shift(UP*3.5)
        Condition = Text('У чотирикутнику ABCD відомо, що ∠A = ∠C = 90\nДоведіть, що бісектриси двох інших кутів чотирикутникуа або паралельні,\nабо лежать на одній прямій').scale(0.5)
        self.play(FadeIn(title))
        self.play(Write(Condition), run_time = 5)
        self.wait(2)
        self.play(FadeOut(Condition))
        
        text_1 = Tex("$1)$").shift(LEFT*6+UP*3)
        self.play(Write(text_1))
        
        #Dots
                
        dot_A = Dot()
        
        dot_B = dot_A.copy()
        dot_B.shift(UP*2+RIGHT*2)
        
        dot_C = dot_B.copy()
        dot_C.shift(RIGHT)
        
        dot_D = dot_A.copy()
        dot_D.shift(DOWN*3+RIGHT*3)
        
        dots = VGroup(*[
            dot_A, dot_B, dot_C, dot_D
        ]).shift(UP+RIGHT*3)
        
        #self.add(dots)
        
        line_AB = Line(dot_A, dot_B)
        line_BC = Line(dot_B, dot_C)
        line_CD = Line(dot_C, dot_D)
        line_DA = Line(dot_D, dot_A)
        
        lines = VGroup(*[
            line_AB, line_BC, line_CD, line_DA
        ])
        
        text_A = Tex("$A$").scale(0.75).next_to(dot_A, LEFT)
        text_B = Tex("$B$").scale(0.75).next_to(dot_B,LEFT)
        text_C = Tex("$C$").scale(0.75).next_to(dot_C, RIGHT)
        text_D = Tex("$D$").scale(0.75).next_to(dot_D,LEFT)
        
        angle_A = RightAngle(Line(dot_A, dot_A.copy().shift(DOWN+RIGHT)),
                            Line(dot_A, dot_A.copy().shift(UP+RIGHT)))
        angle_C = RightAngle(Line(dot_C, dot_C.copy().shift(LEFT))
                           ,Line(dot_C, dot_C.copy().shift(DOWN)))
        
        
        
        self.play(Create(Line(dot_A, dot_A.copy().shift(UP+RIGHT)))
                ,Create(Line(dot_A, dot_A.copy().shift(DOWN+RIGHT)))
                ,Create(angle_A)
                ,Create(Line(dot_C, dot_C.copy().shift(LEFT)))
                ,Create(Line(dot_C, dot_C.copy().shift(DOWN)))
                ,Create(angle_C)
                ,Write(text_C)
                ,Write(text_A)
                ,Create(dot_C)
                ,Create(dot_A)
                )
        self.wait()
        
        self.play(Create(lines)
                 ,Create(dot_B)
                 ,Create(dot_D)
                 ,Write(text_B)
                 ,Write(text_D)
                 )
        self.wait()
        
        #Dot_E
        angle_B = Angle(line_AB, line_BC, radius = 0.4, quadrant=(-1,1))
    
        self.wait()
        
        dot_E = dot_B.copy()
        dot_E.shift((DOWN*math.sin(angle_B.get_value()/2)
                   +RIGHT*math.cos(angle_B.get_value()/2))/(math.cos(angle_B.get_value()/2)))
        line_E = Line(dot_B, dot_E)
        text_E = Tex("$E$").scale(0.75).next_to(dot_E, RIGHT)
        
        self.play(Create(dot_E)
                 ,Create(line_E)
                 ,Write(text_E)
                 )
        
        self.wait()
        
        angle_ABE = Angle(line_AB, line_E, radius = 0.4,quadrant=(-1,1))
        angle_CBE = Angle(line_BC, line_E, radius = 0.3,quadrant=(1,1),other_angle=True)
        
        self.play(Create(angle_ABE)
                 ,Create(angle_CBE)
                 )
        
        self.wait()
        
        #Dot_F
        
        angle_D = Angle(line_DA, line_CD,radius = 0.4,quadrant=(1,-1),other_angle = True)
        
        
        #self.play(Create(angle_D))
        
        dot_F = dot_D.copy()
        dot_F.shift((UP*math.sin(3*PI/8)
                   +LEFT*math.cos(3*PI/8))*4.24/math.cos(PI/8))
        line_F = Line(dot_F, dot_D)
        text_F = Tex("$F$").scale(0.75).next_to(dot_F,LEFT)
        
        #print(line_DA.get_length())
        self.play(Create(dot_F)
                 ,Create(line_F)
                 ,Write(text_F)
                 )
        
        text_dots = VGroup(*[
            text_A, text_B, text_C, text_D, text_E, text_F
        ])
        
        
        angle_ADF_1 = Angle(line_DA, line_F,radius = 0.5, quadrant=(1,-1),other_angle = True)
        angle_ADF_2 = Angle(line_DA, line_F,radius = 0.6, quadrant=(1,-1),other_angle = True)
        angle_CDF_1 = Angle(line_CD, line_F, radius = 0.4, quadrant=(-1,-1))
        angle_CDF_2 = Angle(line_CD, line_F, radius = 0.5, quadrant=(-1,-1))
        
        self.play(Create(angle_CDF_1)
                 ,Create(angle_CDF_2)
                 ,Create(angle_ADF_1)
                 ,Create(angle_ADF_2)
                 )
        
        #
        TEXT1 = Text("∠ABE = ∠CBE = ½∠ABC").scale(0.5).shift(LEFT*4+UP*2.5)
        
        TEXT2 = Text("Нехай ∠CBE = x, а ∠ABC = 2x").scale(0.5).next_to(TEXT1,DOWN)
    
        self.play(Write(TEXT1))
        self.wait()
        self.play(Write(TEXT2))
        self.wait()
        
        #ΔCBE_Title
        
        CBE_text_title = Text("ΔCBE").scale(0.5).set_color(ORANGE)
        CBE_text_C = Text("С = ").scale(0.5).set_color(ORANGE)
        CBE_text_B = Text("B = ").scale(0.5).set_color(ORANGE)
        CBE_text_E = Text("E = ").scale(0.5).set_color(ORANGE)
        CBE_text = VGroup(*[
            CBE_text_title,CBE_text_C,CBE_text_B,CBE_text_E
        ]).arrange(DOWN, buff = 0.3).next_to(TEXT2,DOWN).shift(LEFT*2)
        
        self.play(Write(CBE_text))
        self.wait()
        
        #CBE_angles
        
        CBE_angle_E = Angle(line_CD, line_E, radius = 0.5, quadrant = (-1,-1)).set_color(ORANGE)
        self.play(Create(CBE_angle_E)
                 ,angle_CBE.animate.set_color(ORANGE)
                 ,angle_C.animate.set_color(ORANGE)
                 )
        self.wait()
        
        #ΔCBE_Indicate
        
        CBE = VGroup(*[
            dot_C, dot_B, dot_E,
            Line(dot_C, dot_E), line_BC,line_E,
            angle_CBE, angle_C
        ])
        
        self.play(Indicate(CBE,color = ORANGE))
        self.wait()
        
        
        
        #ΔCBE_values
        
        CBE_value_B = Text("x").scale(0.5).next_to(CBE_text_B,RIGHT).set_color(ORANGE)
        CBE_value_C = Text("90").scale(0.5).next_to(CBE_text_C, RIGHT).set_color(ORANGE)
        CBE_value_E = Text("180 - 90 - x").scale(0.5).next_to(CBE_text_E).set_color(ORANGE)
        CBE_value_E1 = Text("90 - x").scale(0.5).next_to(CBE_text_E).set_color(ORANGE)
        
        CBE_values = VGroup(*[
            CBE_value_B, CBE_value_C, CBE_value_E
        ])
        
        self.play(Transform(angle_CBE.copy(), CBE_value_B)
                 ,Transform(angle_C.copy(), CBE_value_C)
                 ,Write(CBE_value_E)
                 )
        self.wait()
        self.play(ReplacementTransform(CBE_value_E,CBE_value_E1))
        self.wait()
        
        #CBE_finish
        
        self.play(angle_C.animate.set_color(WHITE)
                 ,angle_CBE.animate.set_color(WHITE)
                 ,FadeOut(CBE_angle_E)
                 )
        
        
        #ABCD_title
        
        ABCD_text_title = Text("ABCD").scale(0.5).set_color(YELLOW)
        ABCD_text_A = Text("A = ").scale(0.5).set_color(YELLOW)
        ABCD_text_B = Text("B = ").scale(0.5).set_color(YELLOW)
        ABCD_text_C = Text("C = ").scale(0.5).set_color(YELLOW)
        ABCD_text_D = Text("D = ").scale(0.5).set_color(YELLOW)
        ABCD_text = VGroup(*[
            ABCD_text_title
            ,ABCD_text_A
            ,ABCD_text_B
            ,ABCD_text_C
            ,ABCD_text_D
        ]).arrange(DOWN, buff = 0.3).next_to(TEXT2,DOWN).shift(LEFT*0.2)
        
        self.play(Write(ABCD_text))
        self.wait()
        
        #ABCD_angles
        
        ABCD_angles = VGroup(*[
            angle_A,
            angle_ABE, angle_CBE,
            angle_C,
            angle_CDF_1, angle_CDF_2, angle_ADF_1, angle_ADF_2
        ])
        
        self.play(ABCD_angles.animate.set_color(YELLOW))
        self.wait()
        
        #ABCD_Indicate
        
        ABCD = VGroup(*[
            dot_A, dot_B, dot_C, dot_D
            ,line_AB, line_BC, line_CD, line_DA
            ,angle_A
            ,angle_C
            ,angle_ABE,angle_CBE
            ,angle_ADF_1, angle_ADF_2, angle_CDF_1, angle_CDF_2
        ])
        
        self.play(Indicate(ABCD, color = YELLOW))
        self.wait()
        
        #ABCD_values
        
        ABCD_value_A = Text("90").scale(0.5).set_color(YELLOW).next_to(ABCD_text_A, RIGHT)
        ABCD_value_B = Text("2x").scale(0.5).set_color(YELLOW).next_to(ABCD_text_B, RIGHT)
        ABCD_value_C = Text("90").scale(0.5).set_color(YELLOW).next_to(ABCD_text_C, RIGHT)
        ABCD_value_D = Text("360 - 90 - 90 - 2x").scale(0.5).set_color(YELLOW).next_to(ABCD_text_D, RIGHT)
        ABCD_value_D1 = Text("360 - (90 + 90) - 2x").scale(0.5).set_color(YELLOW).next_to(ABCD_text_D, RIGHT)
        ABCD_value_D2 = Text("360 - 180 - 2x").scale(0.5).set_color(YELLOW).next_to(ABCD_text_D, RIGHT)
        ABCD_value_D3 = Text("180 - 2x").scale(0.5).set_color(YELLOW).next_to(ABCD_text_D, RIGHT)
        
        ABCD_values = VGroup(*[
            ABCD_value_A, ABCD_value_B, ABCD_value_C, ABCD_value_D
        ])
        
        self.play(Transform(angle_A.copy(),ABCD_value_A)
                 ,Transform(Angle(line_AB, line_BC, radius = 0.4, quadrant=(-1,1)), ABCD_value_B)
                 ,Transform(angle_C.copy(),ABCD_value_C)
                 ,Write(ABCD_value_D)
                 )
        self.wait()
        self.play(ReplacementTransform(ABCD_value_D,ABCD_value_D1))
        self.wait(0.5)
        self.play(ReplacementTransform(ABCD_value_D1,ABCD_value_D2))
        self.wait(0.5)
        self.play(ReplacementTransform(ABCD_value_D2,ABCD_value_D3))
        self.wait()
        
        #ABCD_finish
        
        self.play(ABCD_angles.animate.set_color(WHITE))
        
        #TEXT_3
        
        TEXT3 = Text("∠ADF = ∠CDF = ½∠ACD = 90 - x").scale(0.5).shift(3.5*LEFT+DOWN*1.25)
        self.play(Write(TEXT3))
        self.wait()
        
        #ΔFAD_Title
        
        FAD_text_title = Text("ΔFAD").scale(0.5).set_color(RED)
        FAD_text_D = Text("D =").scale(0.5).set_color(RED)
        FAD_text_A = Text("A =").scale(0.5).set_color(RED)
        FAD_text_F = Text("F =").scale(0.5).set_color(RED)
        FAD_text = VGroup(*[
            FAD_text_title,FAD_text_D,FAD_text_A,FAD_text_F
        ]).arrange(DOWN, buff = 0.3).next_to(TEXT2,DOWN).shift(RIGHT*2)
        
        self.play(Write(FAD_text))
        self.wait()
        
        #ΔFAD_angles
        
        FAD_angle_F = Angle(line_AB, line_F, radius = 0.4, quadrant = (-1,1))
        
        FAD_angles = VGroup(*[
            angle_A,
            FAD_angle_F,
            angle_ADF_1, angle_ADF_2
        ])
        
        self.play(Create(FAD_angle_F)
                 ,FAD_angles.animate.set_color(RED)
                 )
        
        #ΔFAD_Indicate
        
        FAD = VGroup(*[
            dot_F, dot_A, dot_D,
            Line(dot_A, dot_F), line_F, line_DA,
            FAD_angles
        ])
        
        self.play(Indicate(FAD, color = RED))
        self.wait()
        
        #ΔFAD_values
        
        FAD_value_D = Text("90 - x").scale(0.5).next_to(FAD_text_D,RIGHT).set_color(RED)
        FAD_value_A = Text("90").scale(0.5).next_to(FAD_text_A, RIGHT).set_color(RED)
        FAD_value_F = Text("180 - 90 - (90 - x)").scale(0.5).next_to(FAD_text_F).set_color(RED)
        FAD_value_F1 = Text("180 - 90 - 90 + x").scale(0.5).next_to(FAD_text_F).set_color(RED)
        FAD_value_F2 = Text("180 - (90 + 90 - x) ").scale(0.5).next_to(FAD_text_F).set_color(RED)
        FAD_value_F3 = Text("180 - 180 + x").scale(0.5).next_to(FAD_text_F).set_color(RED)
        FAD_value_F4 = Text("x").scale(0.5).next_to(FAD_text_F).set_color(RED)
        
        FAD_values = VGroup(*[
            FAD_value_A, FAD_value_D, FAD_value_F
        ])
        
        self.play(Transform(angle_ADF_1.copy(), FAD_value_D)
                 ,Transform(angle_A.copy(), FAD_value_A)
                 ,Write(FAD_value_F)
                 )
        self.wait()
        self.play(ReplacementTransform(FAD_value_F,FAD_value_F1))
        self.wait(0.5)
        self.play(ReplacementTransform(FAD_value_F1,FAD_value_F2))
        self.wait(0.5)
        self.play(ReplacementTransform(FAD_value_F2,FAD_value_F3))
        self.wait(0.5)
        self.play(ReplacementTransform(FAD_value_F3,FAD_value_F4))
        self.wait()
        
        #CBE_finish
        
        self.play(angle_A.animate.set_color(WHITE)
                 ,angle_ADF_1.animate.set_color(WHITE)
                 ,angle_ADF_2.animate.set_color(WHITE)
                 ,FadeOut(FAD_angle_F)
                 )        
        #Result Ch. 1
        
        CBE_res = VGroup(*[
            CBE_text_B,CBE_value_B
        ])
        FAD_res = VGroup(*[
            FAD_text_F,FAD_value_F4
        ])
        self.play(Circumscribe(CBE_res)
                 ,Circumscribe(FAD_res)
                 )
        self.wait()
        
        TEXT4 = Text("Пряма BE || FD, що і треба було довести").scale(0.5).next_to(TEXT3, DOWN)
        
        self.play(Write(TEXT4))
        self.wait()
        
        #Start Ch. 2
        
        #remove_ch.1
        
        text_2 = Tex("$2)$").shift(LEFT*6+UP*3)
        
        all_except_title = VGroup(*[
            dot_A, dot_B, dot_C, dot_D, dot_E, dot_F
           ,line_AB, line_BC, line_CD, line_DA, line_E, line_F
           ,angle_A, angle_ABE, angle_C, angle_ADF_1, angle_ADF_2, angle_CBE
           ,TEXT1, TEXT2, TEXT3, TEXT4
           ,CBE_text, ABCD_text, FAD_text
           ,CBE_values, ABCD_values, FAD_values
           ,text_dots
        ])
        
        #print(self.get_mobject_family_members())
        self.play(FadeOut(*[subm for subm in self.get_mobject_family_members()][23::]))
        self.wait()
        
        
        
        #text_2 = Tex("$2)$").shift(LEFT*6+UP*3)
        self.play(Write(text_2))
        
        dot_A = Dot()
        
        dot_B = dot_A.copy()
        dot_B.shift(UP+2*RIGHT)
        
        dot_C = dot_A.copy()
        dot_C.shift(RIGHT*4)
        
        dot_D = dot_A.copy()
        dot_D.shift(4*DOWN + 2*RIGHT)
        
        dots = VGroup(*[
            dot_A, dot_B, dot_C, dot_D
        ]).shift(RIGHT*2+UP*2.5)
        #self.add(dots)
        
        line_AB = Line(dot_A, dot_B)
        line_BC = Line(dot_B, dot_C)
        line_CD = Line(dot_C, dot_D)
        line_DA = Line(dot_D, dot_A)
        
        lines = VGroup(*[
            line_AB, line_BC, line_CD, line_DA
        ])
        
        text_A = Tex("$A$").scale(0.75).next_to(dot_A, LEFT)
        text_B = Tex("$B$").scale(0.75).next_to(dot_B,LEFT)
        text_C = Tex("$C$").scale(0.75).next_to(dot_C, RIGHT)
        text_D = Tex("$D$").scale(0.75).next_to(dot_D,LEFT)
        
        angle_A = RightAngle(Line(dot_A, dot_A.copy().shift(2*DOWN+RIGHT)),
                            Line(dot_A, dot_A.copy().shift(UP+2*RIGHT)))
        angle_C = RightAngle(Line(dot_C, dot_C.copy().shift(2*LEFT+UP))
                           ,Line(dot_C, dot_C.copy().shift(2*DOWN+LEFT)))
        
        self.play(Create(lines)
                 ,Create(dot_B)
                 ,Create(dot_D)
                 ,Write(text_B)
                 ,Write(text_D)
                 ,Write(text_C)
                 ,Write(text_A)
                 ,Create(dot_C)
                 ,Create(dot_A)
                 )
        
        self.play(Create(angle_A)
                 ,Create(angle_C)
                 )
        self.wait()
        
        #Angles
        
        line_BD = Line(dot_B, dot_D)
        
        angle_ABD = Angle(line_AB, line_BD, radius=0.4, quadrant=(-1,1))
        angle_CBD = Angle(line_BC, line_BD, radius=0.5, quadrant=(1,1), other_angle = True)
        
        TEXT1 = Text("За умовою, BD - бісектриса ∠ABC").scale(0.5).next_to(text_2,DOWN).shift(RIGHT*2)

        
        self.play(Write(TEXT1)
                 ,Create(line_BD)
                 ,Create(angle_ABD)
                 ,Create(angle_CBD)
                 )
        self.wait()
        TEXT2 = Text("∠ABD = ∠CBD, BD - спільна гіпотенуза").scale(0.5).next_to(TEXT1,DOWN).shift(RIGHT*0.5)
        TEXT3 = Text("За ознакою рівності маємо:").scale(0.5).next_to(TEXT2,DOWN).shift(LEFT)
        TEXT4 = Text("ΔABD =").scale(0.5).next_to(TEXT3, RIGHT)
        TEXT5 = Text("ΔCBD").scale(0.5).next_to(TEXT4, RIGHT)
        
        self.play(Write(TEXT2))
        self.wait(0.5)
        self.play(Write(TEXT3))
        self.wait(0.5)
        self.play(Circumscribe(VGroup(*[
            line_AB, line_BD, line_DA
        ]))
                 ,Write(TEXT4))
        self.wait(0.5)
        self.play(Circumscribe(VGroup(*[
            line_BC,line_CD,line_BD
        ]))
                 ,Write(TEXT5))
        self.wait()
        
        
        TEXT6 = Text("За властивістю рівних фігур маємо:").scale(0.5).next_to(TEXT3,DOWN).shift(RIGHT*.6)
        TEXT7 = Text("∠ADB =").scale(0.5).next_to(TEXT6, RIGHT)
        TEXT8 = Text("∠CDB").scale(0.5).next_to(TEXT7, RIGHT)
        TEXT9 = Text("Звідси можна зробити висновок, BD - бісектриса ∠ADC").scale(0.5).next_to(TEXT6,DOWN).shift(RIGHT*1.5)
        
        angle_ADB_1 = Angle(line_DA, line_BD, radius = 0.5,quadrant=(1,-1), other_angle=True)
        angle_ADB_2 = Angle(line_DA, line_BD, radius = 0.6,quadrant=(1,-1), other_angle=True)
        angle_CDB_1 = Angle(line_CD, line_BD, radius= 0.6, quadrant = (-1,-1))
        angle_CDB_2 = Angle(line_CD, line_BD, radius= 0.7, quadrant = (-1,-1))
        
        self.play(Create(angle_ADB_1)
                 ,Create(angle_ADB_2)
                 ,Create(angle_CDB_1)
                 ,Create(angle_CDB_2)
                 )
        self.play(Write(TEXT6))
        self.wait(0.5)
        self.play(Flash(angle_ADB_2)
                 ,Write(TEXT7)
                 )
        self.wait(0.5)
        self.play(Flash(angle_CDB_2)
                 ,Write(TEXT8)
                 )
        self.wait(0.5)
        self.play(Write(TEXT9))
        self.wait()
        
        self.play(FadeOut(*[subm for subm in self.get_mobject_family_members()]))
        
        Goodbye = MarkupText('Дякую за перегляд', gradient=(BLUE, YELLOW))
        
        self.play(Write(Goodbye))
        
        self.wait(1)
        self.play(FadeOut(Goodbye))
```
