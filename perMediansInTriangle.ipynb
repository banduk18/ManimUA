```python
from manim import *
config.media_width = "75%"
config.verbosity = "WARNING"

import math
```

```python
%%manim -qh -v WARNING perMediansInTriangle

class perMediansInTriangle(Scene):
    def construct(self):
        
        #const
        alpha = 0.6
        
        #dots
        dot_A = Dot()
        dot_B = dot_A.copy().shift((UP*math.sin(alpha) + RIGHT*math.cos(alpha))*6)
        dot_C = dot_A.copy().shift(RIGHT*8)
        dot_M = Dot(Line(dot_A, dot_B).get_center())
        dot_N = Dot(Line(dot_A, dot_C).get_center())

        angle_BNC = Angle(Line(dot_B, dot_N), Line(dot_N, dot_C), radius=0.4, quadrant=(-1,1), other_angle=True)
        dot_O = dot_N.copy().shift((DOWN*math.sin(angle_BNC.get_value()) + RIGHT*math.cos(angle_BNC.get_value()))*Line(dot_N, dot_B).get_length()/2.85)
        
        dots = VGroup(*[dot_A, dot_B, dot_C, dot_M, dot_N, dot_O]).shift(LEFT*1.5)
        
        #lines
        line_AB = Line(dot_A, dot_B)
        line_BC = Line(dot_B, dot_C)
        line_CA = Line(dot_C, dot_A)
        line_CM = Line(dot_C, dot_M)
        line_BN = Line(dot_B, dot_N)
        
        lines = VGroup(*[line_AB, line_BC, line_CA, line_CM, line_BN])
        
        #labels
        label_A = Tex(r"A").scale(0.85).next_to(dot_A, DOWN)
        label_B = Tex(r"B").scale(0.85).next_to(dot_B)
        label_C = Tex(r"C").scale(0.85).next_to(dot_C, DOWN)
        label_M = Tex(r"M").scale(0.85).next_to(dot_M, UL)
        label_N = Tex(r"N").scale(0.85).next_to(dot_N, DOWN)
        label_O = Tex(r"O").scale(0.85).next_to(dot_O, UR)
        
        labels = VGroup(*[label_A, label_B, label_C, label_M, label_N, label_O])
        
        #dashes
        #N1
        dot_N1 = Dot(Line(dot_N, dot_A).get_center())
        dash_N1 = Line(dot_N1.copy().shift((DOWN*math.sin(angle_BNC.get_value())+ RIGHT*math.cos(angle_BNC.get_value()))*.3),
                       dot_N1.copy().shift((UP*math.sin(angle_BNC.get_value())+ LEFT*math.cos(angle_BNC.get_value()))*.3)
                      )
        
        dashes_N1 = VGroup(*[dash_N1, dash_N1.copy()]).arrange(LEFT, buff=0).move_to(dot_N1)
        dashes_N2 = dashes_N1.copy().move_to(Dot(Line(dot_N, dot_C).get_center()))
        #M1
        dot_M1 = Dot(Line(dot_M, dot_A).get_center())
        dash_M1 = Line(dot_M1.copy().shift((UP*math.sin(alpha)+ LEFT*math.cos(alpha))*.3),
                       dot_M1.copy().shift((DOWN*math.sin(alpha)+ RIGHT*math.cos(alpha))*.3)
                      )
        dash_M2 = dash_M1.copy().move_to(Dot(Line(dot_M, dot_B).get_center()))
        
        dashes = VGroup(*[dashes_N1, dashes_N2, dash_M1, dash_M2])
        
        angle_BOC = RightAngle(Line(dot_B, dot_O), Line(dot_O, dot_C), length=0.4, quadrant=(1,-1))
        
        #ANIM_1
        self.play(LaggedStart(Create(VGroup(*[dot_A, dot_B, dot_C]))), lag_ratio = 0.5)
        self.wait()
        self.play(LaggedStart(Write(VGroup(*[label_A, label_B, label_C]))), lag_ratio = 0.5)
        self.wait()
        self.play(LaggedStart(Create(VGroup(*[line_AB, line_BC, line_CA]))),lag_ratio = 0.5)
        self.wait()
        self.play(LaggedStart(Create(VGroup(*[line_CM, dot_M])), Write(label_M)), lag_ratio = 0.5)
        self.wait()
        self.play(LaggedStart(Create(VGroup(*[line_BN, dot_N])), Write(label_N), Create(dot_O), Write(label_O)), lag_ratio = 0.5)
        self.wait()
        self.play(Create(angle_BOC))
        self.wait()
        
        
        TEXT1 = Text("За умовою")
        TEXT1_1 = MathTex(r"BN").scale(1.3)
        TEXT1_2 = Text("i")
        TEXT1_3 = MathTex(r"CM").scale(1.3)
        TEXT1_4 = Text(" - медіани")
        ROW1 = VGroup(TEXT1, TEXT1_1, TEXT1_2, TEXT1_3, TEXT1_4).arrange(RIGHT)
        
        TEXT2 = Text("відповідних сторін")
        TEXT2_1 = MathTex(r"AC").scale(1.3)
        TEXT2_2 = Text("i")
        TEXT2_3 = MathTex(r"AB").scale(1.3)
        ROW2 = VGroup(*[TEXT2, TEXT2_1, TEXT2_2, TEXT2_3]).arrange(RIGHT)
        
        TEXT3 = Text("і дорівнюють")
        TEXT3_1 = MathTex(r"6").scale(1.3)
        TEXT3_2 = Text("та")
        TEXT3_3 = MathTex(r"8").scale(1.3)
        ROW3 = VGroup(*[TEXT3, TEXT3_1, TEXT3_2, TEXT3_3]).arrange(RIGHT)
        
        TEXT4 = Text("Нехай")
        TEXT4_1 = MathTex(r"NO").scale(1.3)
        TEXT4_2 = Text("дорівнює")
        TEXT4_3 = MathTex("x").scale(1.3)
        ROW4 = VGroup(*[TEXT4, TEXT4_1, TEXT4_2, TEXT4_3]).arrange(RIGHT)
        
        TEXT5 = Text("Тоді")
        TEXT5_1 = MathTex(r"OB").scale(1.3)
        TEXT5_2 = Text("буде рівна")
        TEXT5_3 = MathTex(r"2x").scale(1.3)
        ROW5 = VGroup(*[TEXT5, TEXT5_1, TEXT5_2, TEXT5_3]).arrange(RIGHT)
        
        ROWS1 = VGroup(*[ROW1,ROW2,ROW3,ROW4,ROW5]).arrange(DOWN, aligned_edge=LEFT, buff = 0.5).scale(0.65).shift(LEFT*3.5+UP*1.5)
        
        TEXT6 = Text("Розглянемо")
        TEXT6_1 = MathTex(r"\Delta NOC, \angle NOC = 90^\circ").scale(1.3).set_color(YELLOW)
        ROW6 = VGroup(*[TEXT6, TEXT6_1]).arrange(RIGHT)
        
        
        TEXT7 = MathTex(r"CN = 4").scale(1.3).set_color(YELLOW)
        TEXT7_1 = MathTex(r"NO = x").scale(1.3).set_color(YELLOW)
        TEXT7_2 = MathTex(r"OC = \sqrt{16 - x^2}").scale(1.3).set_color(YELLOW)
        TEXT7_3 = MathTex(r"\Rightarrow MO = \frac{\sqrt{16 - x^2}}{2}").scale(1.3).set_color(YELLOW)
        ROW7 = VGroup(*[TEXT7_2, TEXT7_3]).arrange(RIGHT)

        ROWS2 = VGroup(*[ROW6, TEXT7, TEXT7_1,ROW7]).arrange(DOWN, aligned_edge=LEFT, buff = 0.5).scale(0.65).shift(LEFT*3.35+DOWN*2.4)
        
        TEXT8 = Text("Тоді в")
        TEXT8_1 = MathTex(r"\Delta MOB, \angle O = 90^\circ").scale(1.3).set_color(RED)
        ROW8 = VGroup(*[TEXT8, TEXT8_1]).arrange(RIGHT)
        TEXT8_2 = MathTex(r"MO = \frac{\sqrt{16 - x^2}}{2}").scale(1.3).set_color(RED)
        TEXT8_3 = MathTex(r"MB = 3").scale(1.3).set_color(RED)
        TEXT8_4 = MathTex(r"OB = \sqrt{9 - \frac{16 - x^2}{4}}").scale(1.3).set_color(RED)
        
        ROWS3 = VGroup(*[ROW8, TEXT8_2, TEXT8_3, TEXT8_4]).arrange(DOWN, aligned_edge=LEFT, buff = 0.5).scale(0.5).shift(RIGHT*2.5+DOWN*2.45)
        
        d = 0.75
        #ANIM_2
        
        self.play(LaggedStart(Write(TEXT1),
                              Write(TEXT1_1),
                              lag_ratio=d))
        self.play(Indicate(line_BN),Circumscribe(TEXT1_1))
        self.wait()
        
        self.play(LaggedStart(Write(TEXT1_2),
                              Write(TEXT1_3),
                              lag_ratio=d))
        self.play(Indicate(line_CM),Circumscribe(TEXT1_3))
        self.play(Write(TEXT1_4))
        self.play(LaggedStart(Create(dashes), lag_ratio=d))
        self.wait()
        
        self.play(LaggedStart(Write(TEXT2),
                              Write(TEXT2_1), 
                              lag_ratio= d))
        self.play(Indicate(line_CA),Circumscribe(TEXT2_1))
        self.wait()
        
        self.play(LaggedStart(Write(TEXT2_2),
                              Write(TEXT2_3), 
                              lag_ratio=d))
        self.play(Indicate(line_AB),Circumscribe(TEXT2_3))
        self.wait()
        
        self.play(LaggedStart(Write(TEXT3),
                              Write(TEXT3_1),
                              Write(TEXT3_2),
                              Write(TEXT3_3), 
                              lag_ratio = d))
        self.wait()
        
        self.play(LaggedStart(Write(TEXT4),
                              Write(TEXT4_1), 
                              lag_ratio=d))
        self.wait()
        line_NO = Line(dot_N, dot_O)
        self.play(Indicate(line_NO),Circumscribe(TEXT4_1))
        self.play(FadeOut(line_NO))
        self.play(LaggedStart(Write(TEXT4_2),
                              Write(TEXT4_3),
                              lag_ratio=d))
        self.wait()
        
        self.play(LaggedStart(Write(TEXT5),
                              Write(TEXT5_1),
                              lag_ratio=d))
        line_OB = Line(dot_O, dot_B)
        self.play(Indicate(line_OB),Circumscribe(TEXT5_1))
        self.play(FadeOut(line_OB))
        self.play(LaggedStart(Write(TEXT5_2), 
                              Write(TEXT5_3),
                              lag_ratio=d))
        self.wait()
        
        line_OC, line_CN, line_NO = Line(dot_O, dot_C), Line(dot_C, dot_N), Line(dot_N, dot_O)
        self.play(LaggedStart(Write(TEXT6),
                              Write(TEXT6_1),
                              lag_ratio=d))
        self.play(LaggedStart(line_OC.animate.set_color(YELLOW),
                              line_CN.animate.set_color(YELLOW),
                              line_NO.animate.set_color(YELLOW),
                              label_N.animate.set_color(YELLOW),
                              label_O.animate.set_color(YELLOW),
                              label_C.animate.set_color(YELLOW),
                              lag_ratio=d))
        self.wait()
        
        self.play(LaggedStart(Write(TEXT7),
                              Write(TEXT7_1),
                              Write(TEXT7_2),
                              Write(TEXT7_3),
                              lag_ratio=d))
        self.wait()
        
        line_OM, line_MB, line_BO = Line(dot_O, dot_M), Line(dot_M, dot_B), Line(dot_B, dot_O)
        self.play(LaggedStart(Write(TEXT8),
                              Write(TEXT8_1)),lag_ratio=d)
        self.play(LaggedStart(line_OM.animate.set_color(RED),
                              line_MB.animate.set_color(RED),
                              line_BO.animate.set_color(RED),
                              label_M.animate.set_color(RED),
                              label_B.animate.set_color(RED),
                              label_O.animate.set_color(ORANGE),
                              lag_ratio=d))
        self.wait()
        
        self.play(LaggedStart(Write(TEXT8_2),
                              Write(TEXT8_3),
                              Write(TEXT8_4), 
                              lag_ratio = d))
        self.wait()
        
        self.play(Circumscribe(VGroup(*[TEXT5_1, TEXT5_2, TEXT5_3])))
        self.play(Circumscribe(TEXT8_4))
        self.wait()
        
        self.play(FadeOut(VGroup(*[TEXT6, TEXT6_1, TEXT7, TEXT7_1, TEXT7_2, TEXT7_3])))
        self.wait()
        
        TEXT9 = MathTex(r"2x = \sqrt{9 - \frac{16 - x^2}{4}}").scale(1.3)
        TEXT9_1 = MathTex(r"4x^2 = \frac{36 - 16 + x^2}{4}").scale(1.3)
        TEXT9_2 = MathTex(r"15x^2 = 20").scale(1.3)
        TEXT9_3 = MathTex(r"x = \frac{2\sqrt{3}}{3}").scale(1.3)
        
        ROWS4 = VGroup(*[TEXT9, TEXT9_1, TEXT9_2, TEXT9_3]).arrange(DOWN, aligned_edge=LEFT).scale(0.65).shift(DOWN*2+LEFT*4.5)
        
        self.play(LaggedStart(Write(TEXT9),
                              Write(TEXT9_1),
                              Write(TEXT9_2),
                              Write(TEXT9_3),
                              lag_ratio=d))
        self.wait()
        
        self.play(FadeOut(VGroup(*[TEXT8, TEXT8_1, TEXT8_2, TEXT8_3, TEXT8_4])))
        self.wait()
        
        TEXT10 = Text("Тоді в")
        TEXT10_1 = MathTex(r"\Delta BOC \angle BOC = 90^\circ").scale(1.3).set_color(BLUE)
        ROW5 = VGroup(*[TEXT10, TEXT10_1]).arrange(RIGHT)
        TEXT10_2 = MathTex(r"OC = \sqrt{16 - x^2} = \frac{2\sqrt{13}}{3}").scale(1.3).set_color(BLUE)
        TEXT10_3 = MathTex(r"BO = 4\sqrt{3}").scale(1.3).set_color(BLUE)
        TEXT10_4 = MathTex(r"CB = \sqrt{\frac{132+48}{9}} = 2\sqrt{5}").set_color(BLUE)
        
        ROWS5 = VGroup(*[ROW5, TEXT10_2, TEXT10_3, TEXT10_4]).arrange(DOWN, aligned_edge=LEFT).scale(0.65).shift(DOWN*2.25)
        
        self.play(LaggedStart(line_BO.animate.set_color(BLUE),
                              line_BC.animate.set_color(BLUE),
                              line_OC.animate.set_color(BLUE),
                              label_B.animate.set_color(BLUE),
                              label_O.animate.set_color(BLUE),
                              label_C.animate.set_color(BLUE),
                              lag_ratio = d))
        
        self.play(LaggedStart(Write(ROW5),
                              Write(TEXT10_2),
                              Write(TEXT10_3),
                              Write(TEXT10_4),
                              lag_ratio = d))
        self.wait(2)
        
        TEXT11 = MathTex(r"P = 14+2\sqrt{5}").scale(.85).shift(RIGHT*4+DOWN*2.5)
        
        self.play(Write(TEXT11))
        self.wait()
        
        self.play(FadeOut(*[subm for subm in self.get_mobject_family_members()]))
        
        Goodbye = MarkupText('Дякую за перегляд', gradient=(BLUE, YELLOW))
        
        self.play(Write(Goodbye))
        
        self.wait(1)
        self.play(FadeOut(Goodbye))
```
